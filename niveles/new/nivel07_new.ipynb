## Reto 601: Validación de Subconjuntos
* Subset Validation
* Escribe una función que devuelva `True` si todos los subconjuntos en una lista pertenecen a un conjunto dado.

* **Ejemplos**

```
validate_subsets([[1, 2], [2, 3], [1, 3]], [1, 2, 3]) ➞ True
validate_subsets([[1, 2, 3], [2], [3], []], [1, 2, 3]) ➞ True
validate_subsets([[1, 2], [2, 3], [1, 4]], [1, 2, 3]) ➞ False
validate_subsets([[1, 2, 3, 4]], [1, 2, 3]) ➞ False
```

* **Notas**
    - El **conjunto vacío** y el **conjunto** mismo son **ambos** subconjuntos válidos de un conjunto (no estamos hablando de subconjuntos propios aquí).
    - El conjunto y el subconjunto tendrán cada uno elementos únicos.


`````python
# Método 1
def validate_subsets(subsets: list, full_set: list) -> bool:
    return all(num in full_set for subset in subsets for num in subset)
`````

`````python
# Método 2
def validate_subsets(subsets: list, full_set: list) -> bool:
    full_set = set(full_set)  # Convertimos la lista a un conjunto para operaciones más eficientes

    for subset in subsets:
        if not set(subset).issubset(full_set):
            return False

    return True
`````

`````python
# Método 3
def validate_subsets(subsets: list, full_set: list) -> bool:
    full_set = set(full_set)
    return all(set(subset) - full_set == set() for subset in subsets)
`````

`````python
print(validate_subsets([[1, 2], [2, 3], [1, 3]], [1, 2, 3]))    # True
print(validate_subsets([[1, 2, 3], [2], [3], []], [1, 2, 3]))   # True
print(validate_subsets([], [1, 2, 3]))                          # True
print(validate_subsets([[]], []))                               # True
print(validate_subsets([[1]], []))                              # False
print(validate_subsets([[1, 2], [2, 3], [1, 4]], [1, 2, 3]))    # False
print(validate_subsets([[1, 2, 3, 4]], [1, 2, 3]))              # False
`````

## Reto 602: Eliminar Caracteres Repetidos
* Remove Repeated Characters
* Crea una función que elimine cualquier carácter repetido en una palabra pasada a la función.
* No solo caracteres consecutivos, sino caracteres que se repitan en cualquier parte de la entrada.

* **Ejemplos**

```
unrepeated("banana") ➞ "ban"
unrepeated("aaaaa") ➞ "a"
unrepeated("WWE!!!") ➞ "WE!"
unrepeated("llama 112") ➞ "lama 12"
```

* **Notas**
    - No se pasarán más de dos palabras en las pruebas.
    - La entrada incluye caracteres especiales y números.

`````python
# Método 1
def unrepeated(s: str) -> str:
    result = ""
    for c in s:
        if c not in result:
            result += c
    return result
`````

`````python
# Método 2
def unrepeated(s: str) -> str:
    return ''.join(c for i, c in enumerate(s) if c not in s[:i])
`````

`````python
# Método 3: usando un diccionario
# a partir de Python 3.7, los diccionarios mantienen el orden de inserción por defecto.
def sin_repetir(texto):
    return ''.join(dict.fromkeys(texto))
`````

`````python
print(unrepeated("banana"))     # ban
print(unrepeated("aaaaa"))      # a
print(unrepeated("WWE!!!"))     # WE!
print(unrepeated("llama 112"))  # lam 12
`````

#### Explicación del Método 6
- **`filter`**: Esta función aplica una condición a cada elemento en `word`. Solo los elementos que cumplen la condición se mantienen.
- **`lambda x, seen=set(): not (x in seen or seen.add(x))`**:
  - **`seen=set()`**: Se utiliza un conjunto `seen` para rastrear los caracteres que ya han aparecido.
  - **`x in seen or seen.add(x)`**: Este truco inteligente agrega el carácter a `seen` si no está presente y devuelve `True` si el carácter ya estaba, lo que causa que `filter` lo excluya.
  - **`not (x in seen or seen.add(x))`**: Negamos la expresión para incluir solo caracteres que no han sido vistos antes.

`````python
def first_unique_letters(words):
    seen = set()
    return [word[0] for word in words if not (word[0] in seen or seen.add(word[0]))]

# Ejemplo de uso
words = ["apple", "banana", "cherry", "avocado", "blueberry", "carrot"]
print(first_unique_letters(words))  # ➞ ['a', 'b', 'c']
`````

##### Otro ejemplo de este "truco" usando una lista de números

`````python
def first_unique_numbers(numbers):
    seen = set()
    return [x for x in numbers if not (x in seen or seen.add(x))]

# Ejemplo de uso
numbers = [4, 9, 6, 4, 7, 5, 8, 6, 9, 5]
print(first_unique_numbers(numbers))        # 4, 9, 6, 7, 5, 8]
`````

## Reto 604: Palabras Estiradas
* Stretched Words
* Escribe una función que tome una cadena y devuelva una nueva cadena con cualquier letra duplicada *consecutiva* eliminada.

* **Ejemplos**

```
desestira("ppoeemm") ➞ "poem"
desestira("wiiiinnnnd") ➞ "wind"
desestira("ttiiitllleeee") ➞ "title"
desestira("cccccaaarrrbbonnnnn") ➞ "carbon"
```

* **Notas**
    - Las cadenas finales *no* incluirán palabras con letras dobles (por ejemplo, "passing", "lottery").

`````python
import re
`````

`````python
# Método 1
def desestira(s: str) -> str:
    if len(s) <= 1:
        return s
    result = s[0]
    for i in range(1, len(s)):
        if s[i-1] != s[i]:
            result += s[i]
    return result
`````

`````python
# Método 2
def desestira(palabra: str) -> str:
    resultado = palabra[0]  # Comenzamos con la primera letra
    for letra in palabra[1:]:  # Iteramos desde la segunda letra
        if letra != resultado[-1]:  # Si la letra actual es diferente a la última en el resultado
            resultado += letra  # La añadimos al resultado
    return resultado
`````

`````python
# Método 3
def desestira(palabra: str) -> str:
    return re.sub(r'(.)\1+', r'\1', palabra)
`````

`````python
print(desestira("ppoeemm"))             # "poem"
print(desestira("wiiiinnnnd"))          # "wind"
print(desestira("ttiiitllleeee"))       # "title"
print(desestira("cccccaaarrrbbonnnnn")) # "carbon"
`````

* **Ejemplos**

```
eval_factorial(["2!", "3!"]) ➞ 8

eval_factorial(["5!", "4!", "2!"]) ➞ 146

eval_factorial(["0!", "1!"]) ➞ 2
```

* **Notas**
    - 0! y 1! ambos son iguales a 1.

`````python
import math
`````

`````python
# Método 1

def eval_factorial(lst: list[str]) -> int:
    return sum(math.factorial(int(num[:-1])) for num in lst)
`````

`````python
# Método 2
def eval_factorial(lst: list[str]) -> int:
    def factorial(n):
        if n == 0 or n == 1:
            return 1
        return n * factorial(n - 1)

    total = 0
    for expr in lst:
        num = int(expr[:-1])  # Elimina el '!' y convierte a entero
        total += factorial(num)

    return total
`````

`````python
# Método 3: usando replace y un bucle for para calcular el factorial
def eval_factorial(lst: list[str]) -> int:
    def factorial(n):
        if n == 0 or n == 1:
            return 1
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

    total = 0
    for expr in lst:
        num = int(expr.replace('!', ''))
        total += factorial(num)

    return total
`````

`````python
print(eval_factorial(["2!", "3!"]))         # 8
print(eval_factorial(["5!", "4!", "2!"]))   # 146
print(eval_factorial(["0!", "1!"]))         # 2
`````

## Reto 606: Lista Balanceada
* Balanced List
* Dada una lista de longitud **par**, copia la mitad con la suma más alta de números a la otra mitad de la lista.
* Si la suma de la primera mitad es igual a la suma de la segunda mitad, devuelve la lista original.
* **Ejemplos**

```
balanced([1, 2, 4, 6, 3, 1]) ➞ [6, 3, 1, 6, 3, 1]
#1 + 2 + 4 < 6 + 3 + 1

balanced([88, 3, 27, 5, 9, 0, 13, 10]) ➞ [88, 3, 27, 5, 88, 3, 27, 5]
#88 + 3 + 27 + 5 > 9 + 0 + 13 + 10

balanced([7, 5, 2, 6, 1, 0, 1, 5, 2, 7, 0, 6]) ➞ [7, 5, 2, 6, 1, 0, 1, 5, 2, 7, 0, 6]
#7 + 5 + 2 + 6 + 1 + 0 = 1 + 5 + 2 + 7 + 0 + 6
```

* **Notas**
    - La longitud de la lista es **par**.

`````python
# Método 1
def balanced(lst: list[int]) -> list[int]:
    n = len(lst)
    mid = n // 2

    first_half = lst[:mid]
    second_half = lst[mid:]

    sum_first = sum(first_half)
    sum_second = sum(second_half)

    if sum_first > sum_second:
        return first_half * 2
    elif sum_second > sum_first:
        return second_half * 2
    else:
        return lst
`````

`````python
# Método 2
def balanced(lst: list[int]) -> list[int]:
    mid = len(lst)//2
    izquierda = lst[:mid]
    derecha = lst[mid:]
    if sum(izquierda) > sum(derecha):
        return izquierda *2
    elif sum(derecha) > sum(izquierda):
        return derecha * 2
    else:
        return lst
`````

`````python
# Método 3
def balanced(lst: list[int]) -> list[int]:
    mid = len(lst) // 2
    left, right = lst[:mid], lst[mid:]
    return lst if sum(left) == sum(right) else (right * 2 if sum(right) > sum(left) else left * 2)
`````

`````python
print(balanced([1, 2, 4, 6, 3, 1]))                     # [6, 3, 1, 6, 3, 1]
print(balanced([88, 3, 27, 5, 9, 0, 13, 10]))           # [88, 3, 27, 5, 88, 3, 27, 5]
print(balanced([7, 5, 2, 6, 1, 0, 1, 5, 2, 7, 0, 6]))   # [7, 5, 2, 6, 1, 0, 1, 5, 2, 7, 0, 6]
print(balanced([10, 5, 15, 20]))                        # [15, 20, 15, 20]
print(balanced([3, 2, 1, 1, 2, 3]))                     # [3, 2, 1, 1, 2, 3]
print(balanced([10, 20, 30, 6, 7, 8]))                  # [10, 20, 30, 10, 20, 30]
`````

    - Todas las entradas proporcionadas en los escenarios de prueba son válidas.
    - Si `start_of_range` es mayor que `end_of_range`, devuelve un array ordenado en orden **descendente**, de lo contrario, ordenado en orden **ascendente**.

`````python
# Método 1
def reversible_inclusive_list(start_of_range: int, end_of_range: int) -> list[int]:
    if start_of_range <= end_of_range:
        return list(range(start_of_range, end_of_range + 1))
    else:
        return list(range(start_of_range, end_of_range - 1, -1))
`````

`````python
# Método 2
def reversible_inclusive_list(start_of_range: int, end_of_range: int) -> list[int]:
    step = 1 if start_of_range < end_of_range else -1
    return [x for x in range(start_of_range, end_of_range + step, step)]
`````

## Reto 608: Mini Sudoku
* Mini Sudoku
* Un Sudoku es una cuadrícula de 9x9 que se completa cuando cada cuadrado de 3x3, fila y columna contiene los números del 1 al 9.
* Para esta tarea, se te dará un cuadrado de 3x3 completo, en forma de una lista bidimensional. Crea una función que verifique que este cuadrado de 3x3 contiene cada número del 1 al 9 exactamente una vez. Asegúrate de que no haya duplicados ni números fuera de este rango.

* **Ejemplos**

```
es_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]) ➞ True

es_mini_sudoku([[8, 9, 2], [5, 6, 1], [3, 7, 4]]) ➞ True

es_mini_sudoku([[1, 1, 3], [6, 5, 4], [8, 7, 9]]) ➞ False
#El 1 se repite dos veces

es_mini_sudoku([[0, 1, 2], [6, 4, 5], [9, 8, 7]]) ➞ False
#El 0 está incluido (fuera del rango)
```

`````python
# Método 1
def es_mini_sudoku(matrix: list[list[int]]) -> bool:
    lista = [n for row in matrix for n in row]
    return sorted(lista) == list(range(1, 10))
`````

`````python
# Método 2
def es_mini_sudoku(cuadrado: list[list[int]]) -> bool:
    # Aplanar la lista bidimensional
    numeros = [num for fila in cuadrado for num in fila]

    # Verificar si hay 9 números únicos en el rango 1-9
    return set(numeros) == set(range(1, 10))
`````

`````python
# Método 3
def es_mini_sudoku(cuadrado: list[list[int]]) -> bool:
    # Convertir la lista de listas en una lista plana usando el operador *
    numeros = [*sum(cuadrado, [])]

    # Verificar que todos los números estén en el rango 1-9
    if not all(1 <= num <= 9 for num in numeros):
        return False

    # Verificar que no haya duplicados
    return len(numeros) == len(set(numeros)) == 9
`````

`````python
print(es_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]))    # True
print(es_mini_sudoku([[8, 9, 2], [5, 6, 1], [3, 7, 4]]))    # True
print(es_mini_sudoku([[1, 1, 3], [6, 5, 4], [8, 7, 9]]))    # False
print(es_mini_sudoku([[0, 1, 2], [6, 4, 5], [9, 8, 7]]))    # False
`````

## Reto 609: Temperaturas Récord
* Record Temperatures
* Se te proporcionan dos listas que contienen datos que representan las temperaturas mínimas y máximas del clima para cada día de la semana.
* La lista de récords contiene las temperaturas récord más bajas/altas de todos los tiempos para ese día de la semana.

```
[[récord mínimo, récord máximo], ...]
```

* La lista de la semana actual contiene las temperaturas mínimas/máximas diarias para cada día de la semana actual.

```
[[mínima diaria, máxima diaria], ...]
```

* Se considera que una temperatura máxima diaria es un nuevo récord máximo si es superior al récord máximo para ese día de la semana. Se considera que una temperatura mínima diaria es un nuevo récord mínimo si es inferior al récord mínimo para ese día de la semana.
* Compara las temperaturas mínimas/máximas diarias de la semana actual con los récords mínimos/máximos y devuelve una lista con las temperaturas récord actualizadas.
* Puede haber múltiples temperaturas récord en una semana.
* Si no se rompen récords, devuelve la lista de récords original.

* **Ejemplo**

```
#             dom       lun      mar        mié      jue       vie       sáb
record_temps([[34, 82], [24, 82], [20, 89],  [5, 88],  [9, 88], [26, 89], [27, 83]],
            [[44, 72], [19, 70], [40, 69], [39, 68], [33, 64], [36, 70], [38, 69]])

➞           [[34, 82], [19, 82], [20, 89], [5, 88], [9, 88], [26, 89], [27, 83]]
```

El récord mínimo anterior para el lunes era 24. La mínima de la semana actual para el lunes fue 19. Por lo tanto, 19 reemplaza a 24 como el nuevo récord mínimo del lunes.

* **Notas**
    - El índice 0 siempre será el mínimo y el índice 1 siempre será el máximo `[mínimo, máximo]`.
    - Como referencia, estas temperaturas están en °F, pero no necesitas convertir ninguna temperatura.

`````python
# Método 1
def record_temps(records: list[list[int]], current_week: list[list[int]]) -> list[list[int]]:
    # Iteramos sobre los días de la semana (0-6)
    for i in range(7):
        # Comparamos la temperatura mínima actual con el récord mínimo
        if current_week[i][0] < records[i][0]:
            records[i][0] = current_week[i][0]

        # Comparamos la temperatura máxima actual con el récord máximo
        if current_week[i][1] > records[i][1]:
            records[i][1] = current_week[i][1]

    return records
`````

`````python
# Método 2: usando min y max.
def record_temps(records: list[list[int]], current_week: list[list[int]]) -> list[list[int]]:
    new_records = []
    for i in range(len(records)):
        new_min = min(records[i][0], current_week[i][0])
        new_max = max(records[i][1], current_week[i][1])
        new_records.append([new_min, new_max])
    return new_records
`````

`````python
# Método 2
def find_a_seat(n: int, tren: list[int]) -> int:
    capacidad = n // len(tren)
    return next((i for i, vagon in enumerate(tren) if vagon <= 0.5 * capacidad), -1)

# next() aplicado a un generador retorna el primer índice que cumple la condición.
# Si no se encuentra, next() retorna -1.
`````

`````python
# Método 3: usando index primero necesitamos filtrar la lista
def find_a_seat(n: int, tren: list[int]) -> int:
    capacidad = n // len(tren)

    # Crear una lista con los valores que cumplen la condición
    filtered = [count for count in tren if count <= 0.5 * capacidad]

    # Si no hay ningún valor que cumpla la condición, devolvemos -1
    if not filtered:
        return -1

    # Usar index para encontrar la primera ocurrencia
    return tren.index(filtered[0])
`````

`````python
# Método 1
def is_circle_collision(circle1: list[int], circle2: list[int]) -> bool:
    r1, x1, y1 = circle1
    r2, x2, y2 = circle2
    distancia = ((x1 - x2) ** 2 + (y1 - y2) ** 2 ) ** .5
    return r1 + r2 > distancia
`````

`````python
# Método 2
def is_circle_collision(circle1: list[int], circle2: list[int]) -> bool:
    radius1, x1, y1 = circle1
    radius2, x2, y2 = circle2

    # Calcula la distancia entre los dos centros de los círculos
    distance_between_centers = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # Compara la distancia con la suma de los radios
    return distance_between_centers < (radius1 + radius2)
`````

```
percentage_changed("$800", "$600") ➞ "25% decrease"
percentage_changed("$1000", "$840") ➞ "16% decrease"
percentage_changed("$100", "$950") ➞ "850% increase"
percentage_changed("$100", "$100") ➞ "0% no change"
```

`````python
# Método 1
def percentage_changed(old: str, new: str) -> str:
    old = int(old[1:])
    new = int(new[1:])
    var = new / old - 1     # variación en tanto por uno
    return f'{var:.0%} increase' if var > 0 else f'{-var:.0%} decrease' if var != 0 else '0% no change'
`````

`````python
# Método 2
def percentage_changed(old: str, new: str) -> str:
    # Elimina los símbolos de dólar y convierte a enteros
    old_price = int(old.replace('$', ''))
    new_price = int(new.replace('$', ''))

    # Si los precios son iguales, no hay cambio
    if old_price == new_price:
        return "0% no change"

    # Calcula el porcentaje de cambio
    change = ((new_price - old_price) / old_price) * 100

    # Redondea al entero más cercano
    change = round(change)

    # Determina si es un aumento o una disminución
    if change > 0:
        return f"{change}% increase"
    else:
        return f"{abs(change)}% decrease"
`````

`````python
# Método 1
def make_happy(frase: str) -> str:
    frase = frase.replace(":(", ":)")
    frase = frase.replace("8(", "8)")
    frase = frase.replace("x(", "x)")
    frase = frase.replace(";(", ";)")
    return frase
`````

`````python
# Método 2
def make_happy(frase: str) -> str:
    frase = list(frase)
    for i in range(1, len(frase)):
        if frase[i] == "(" and frase[i-1] in ":8x;":
            frase[i] = ")"
    return ''.join(frase)
`````

`````python
# Método 3
def make_happy(frase: str) -> str:
    caras_tristes = [':(', '8(', 'x(', ';(']
    caras_felices = [':)', '8)', 'x)', ';)']
    resultado = []
    i = 0
    while i < len(frase):
        # Comprobar si hay una cara triste a partir del carácter actual
        if i < len(frase) - 1 and frase[i:i+2] in caras_tristes:
            # Encontrar el índice correspondiente en caras_tristes y reemplazar por la cara feliz
            indice_cara = caras_tristes.index(frase[i:i+2])
            resultado.append(caras_felices[indice_cara])
            i += 2  # Saltar la cara completa
        else:
            resultado.append(frase[i])
            i += 1
    return ''.join(resultado)
`````

## Reto 624: Mayor Brecha
* Largest Gap
* Dada una lista de números enteros, devuelve la mayor brecha entre elementos de la versión ordenada de esa lista.
* Ejemplo ilustrativo. Considera la lista:

```
[9, 4, 26, 26, 0, 0, 5, 20, 6, 25, 5]
```

... que, después de ordenarla, se convierte en la lista:

```
[0, 0, 4, 5, 5, 6, 9, 20, 25, 26, 26]
```

... de modo que ahora vemos que la mayor brecha en la lista es la brecha de `11` entre 9 y 20.

* **Ejemplos**

```
largest_gap([9, 4, 26, 26, 0, 0, 5, 20, 6, 25, 5]) ➞ 11

largest_gap([14, 13, 7, 1, 4, 12, 3, 7, 7, 12, 11, 5, 7]) ➞ 4

largest_gap([13, 3, 8, 5, 5, 2, 13, 6, 14, 2, 11, 4, 10, 8, 1, 9]) ➞ 2
```

`````python
# Método 1
def largest_gap(lista: list[int]) -> int:
    lista.sort()
    gaps = [lista[i] - lista[i-1] for i in range(1, len(lista))]
    return max(gaps)
`````

`````python
# Método 2
def largest_gap(lista: list[int]) -> int:
    # Ordenamos la lista
    lista_ordenada = sorted(lista)

    # Inicializamos la brecha máxima
    max_brecha = 0

    # Iteramos a través de la lista ordenada
    for i in range(1, len(lista_ordenada)):
        # Calculamos la brecha actual
        brecha_actual = lista_ordenada[i] - lista_ordenada[i-1]

        # Actualizamos la brecha máxima si la actual es mayor
        if brecha_actual > max_brecha:
            max_brecha = brecha_actual

    return max_brecha
`````

`````python
# Método 3
def largest_gap(lista: list[int]) -> int:
    # Implementación de ordenamiento por selección
    for i in range(len(lista)):
        min_idx = i
        for j in range(i+1, len(lista)):
            if lista[j] < lista[min_idx]:
                min_idx = j
        lista[i], lista[min_idx] = lista[min_idx], lista[i]

    # Encontrar la brecha más grande
    max_brecha = 0
    for i in range(1, len(lista)):
        brecha_actual = lista[i] - lista[i-1]
        if brecha_actual > max_brecha:
            max_brecha = brecha_actual

    return max_brecha
`````

## Reto 629: Desviación Estándar
* Standard Deviation
* Las *medidas de tendencia central* (*media*, *moda* y *mediana*) a veces no son suficientes como descriptores en el análisis de un conjunto de datos. Por ejemplo, dadas dos listas `A=[15, 15, 15, 14, 16]` y `B=[2, 7, 14, 22, 30]`, la media es `μ=15` en ambos casos, sin embargo, los valores de la segunda lista están claramente más dispersos del valor promedio.
* La **desviación estándar** (también llamada ***sigma***, la letra griega minúscula ***σ***) **mide la dispersión de los valores en un conjunto de datos** y transforma la afirmación "claramente más disperso que" en una afirmación estadística probada.
La desviación estándar se calcula siguiendo cinco pasos:
1. Obtener la media del conjunto de datos.
2. Para cada valor en el conjunto, calcular la diferencia entre el valor y la media.
3. Elevar al cuadrado cada valor obtenido y sumarlos acumulativamente.
4. Dividir el resultado por la longitud del conjunto de datos.
5. Obtener la raíz cuadrada del valor obtenido.
Dada una lista de valores, devuelve la desviación estándar redondeada a la centésima más cercana.
* **Ejemplos**

```
standard_deviation([3, 5, 7]) ➞ 1.63

standard_deviation([5, 5, 5]) ➞ 0

standard_deviation([-3, -5, -7]) ➞ 1.63
```

* * **Notas**
    - Todas las listas proporcionadas son válidas, no hay excepciones que manejar.
    - Las listas pueden contener enteros positivos o negativos.
    - Recuerda redondear a la centésima más cercana al final.

`````python
import statistics
import statistics as st
from math import sqrt
`````

`````python
# Método 1
def standard_deviation(lst: list[int]) -> float:
    n = len(lst)
    media = sum(lst) / n
    varianza = sum((x - media) ** 2 for x in lst) / n
    return round(varianza ** .5, 2)
`````

`````python
# Método 2

def desviacion_estandar(lista):
    # Calcula la desviación estándar usando statistics.stdev
    desviacion = statistics.stdev(lista)

    # Redondea el resultado a dos decimales
    return round(desviacion, 2)
`````

`````python
# Método 3
def desviacion_estandar(lista):
    # Calcula la desviación estándar usando statistics.pstdev
    desviacion = st.pstdev(lista)

    # Redondea el resultado a dos decimales
    return round(desviacion, 2)
`````

`````python
print(standard_deviation([3, 5, 7]))        # 1.63
print(standard_deviation([5, 5, 5]))        # 0
print(standard_deviation([-3, -5, -7]))     # 1.63
`````

```
power_of_two(32) ➞ True

power_of_two(1) ➞ True

```

`````python
from math import log2
import math
`````

`````python
# Método 1: usando el logaritmo en base 2

def power_of_two(num: int) -> bool:
    return log2(num) - int(log2(num)) < 1e-10
`````

`````python
# Método 2: similar al método anterior
def power_of_two(n: int) -> bool:
    # Manejo de casos especiales
    if n <= 0:
        return False

    # Calculamos el logaritmo en base 2 de n
    log_base_2 = math.log2(n)

    # Verificamos si el logaritmo es un número entero
    return log_base_2.is_integer()
`````

`````python
# Método 3
def power_of_two(n: int) -> bool:
    # Manejo de casos especiales
    if n <= 0:
        return False
    if n == 1:
        return True

    # Dividimos repetidamente por 2
    while n > 1:
        if n % 2 != 0:
            return False
        n = n // 2

    return True
`````

## Reto 632: Unión e Intersección de Listas
* Union and Intersection of Lists
* Crea una función que reciba dos listas y devuelva una lista de intersección y una lista de unión.

- **Lista de Intersección**: Elementos compartidos por ambas listas.
- **Lista de Unión**: Elementos que existen en la primera o segunda lista, o en ambas (no es un OR exclusivo).
- Aunque las listas de entrada puedan tener números duplicados, las listas de intersección y unión devueltas deben estar "conjuntificadas", es decir, sin duplicados. Las listas devueltas deben estar ordenadas en orden ascendente.

* **Ejemplo**

```python
Lista 1: [5, 6, 6, 6, 8, 9]
Lista 2: [3, 3, 4, 4, 5, 5, 8]

Intersección: [5, 8]
# 5 y 8 son los únicos números que existen en ambas listas.

Unión: [3, 4, 5, 6, 8, 9]
# Cada número existe en al menos una lista.
```

**Ejemplos**

```python
intersection_union([1, 2, 3, 4, 4], [4, 5, 9]) ➞ [[4], [1, 2, 3, 4, 5, 9]]

intersection_union([1, 2, 3], [4, 5, 6]) ➞ [[], [1, 2, 3, 4, 5, 6]]

intersection_union([1, 1], [1, 1, 1, 1]) ➞ [[1], [1]]
```

* **Notas**
- El orden de salida debe ser: [Intersection], [Union].
- Recuerda que ambas listas de salida deben estar en orden ascendente.
- Cada lista de entrada tendrá al menos un elemento.
- Si ambas listas son disjuntas (no comparten nada en común), devuelve una lista vacía `[]` para la intersección.

`````python
# Método 1
def intersection_union(list1: list[int], list2: list[int]) -> list:
    # Convertir ambas listas a conjuntos para eliminar duplicados y hacer operaciones de conjuntos
    set1 = set(list1)
    set2 = set(list2)

    # Intersección: elementos comunes a ambos conjuntos
    intersection = sorted(list(set1 & set2))

    # Unión: elementos que están en cualquiera de los dos conjuntos
    union = sorted(list(set1 | set2))

    # Devolver las listas de intersección y unión
    return [intersection, union]
`````

`````python
# Método 2
def intersection_union(list1: list[int], list2: list[int]) -> list:
    union = list(set(list1 + list2))
    intersection = [e for e in union if e in list1 and e in list2]
    return [sorted(intersection), sorted(union)]
`````

`````python
# Método 3
def intersection_union(list1: list[int], list2: list[int]) -> list:
    # Usamos conjuntos para la unión y la intersección
    set1, set2 = set(list1), set(list2)

    # Intersección con filter: filtrar elementos de set1 que estén en set2
    intersection = sorted(filter(lambda x: x in set2, set1))

    # Unión: combinamos ambos conjuntos y los ordenamos
    union = sorted(set1.union(set2))

    return [intersection, union]
`````

## Reto 634: Conjetura de Collatz
* Collatz Conjecture
* Una secuencia de Collatz se genera así. Comience con un número positivo. Si es par, divídalo a la mitad. Si es impar, multiplícalo por tres y suma uno. Repite el proceso con el número resultante.
* La conjetura de Collatz es que cada secuencia eventualmente llega a 1 (continuar más allá de 1 solo resulta en una repetición interminable de la secuencia 4, 2, 1).
* La longitud de la secuencia desde el número inicial hasta el 1 varía ampliamente.
* Cree una función que tome un número como argumento y devuelva una tupla de dos elementos: el número de pasos en la secuencia de Collatz del número y el número más alto alcanzado.

* **Ejemplos**

````
collatz(2) ➞ (2, 2)
#seq = [2, 1]

collatz(3) ➞ (8, 16)
#seq = [3, 10, 5, 16, 8, 4, 2, 1]

collatz(7) ➞ (17, 52)
#seq = [7, 22, 11, 34, 17, 52, 26, 13, 40, 20, 10, 5, 16, 8, 4, 2, 1]

collatz(8) ➞ (4, 8)
#seq = [8, 4, 2, 1]
````

* **Nota**
    - [NO Podrás Resolver este Simple Problema Matemático ¿O Sí?](https://youtu.be/q_dvxXc7d2Y?si=GBRmKCMte4ZR_hzG)

`````python
# Método 1
def collatz(n: int) -> tuple:
    sucesion = [n]
    while n > 1:
        if n % 2:
            n = 3 * n + 1
        else:
            n //= 2
        sucesion.append(n)
    return len(sucesion), max(sucesion)
`````

`````python
# Método 2: recursividad
def collatz_recursive(n: int, steps: int = 0, max_value: int = 0) -> tuple:
    # Caso base: si n es 1, devolvemos los resultados
    if n == 1:
        return steps + 1, max(n, max_value)

    # Actualizamos el valor máximo
    max_value = max(n, max_value)

    # Aplicamos la regla de Collatz
    if n % 2 == 0:
        return collatz_recursive(n // 2, steps + 1, max_value)
    else:
        return collatz_recursive(3 * n + 1, steps + 1, max_value)
`````

`````python
# Método 3
def collatz(n: int) -> tuple:
    return collatz_recursive(n)
`````

`````python
print(collatz(2))       # (2, 2)
print(collatz(3))       # (8, 16)
print(collatz(4))       # (3, 4)
print(collatz(7))       # (17, 52)
print(collatz(8))       # (4, 8)
print(collatz(27))      # (112, 9232)
print(collatz(9663))    # (185, 27114424)
print(collatz(2 ** 60)) # (61, 1152921504606846976)
`````

`````python
# Método 2
def accumulating_product(lst: list) -> list:
    if not lst:  # Si la lista está vacía, devolvemos una lista vacía
        return []

    result = [lst[0]]  # Inicializamos la lista con el primer elemento
    for i in range(1, len(lst)):
        result.append(result[-1] * lst[i])  # Multiplicamos el último valor acumulado por el siguiente valor de la lista
    return result
`````

* The Fibonacci Number
* Cree una función que, dado un número, devuelva el valor correspondiente de ese índice en la serie de Fibonacci.
* La Secuencia de Fibonacci es la serie de números:

`1, 1, 2, 3, 5, 8, 13, 21, 34, ...`

* El siguiente número se encuentra sumando los dos números anteriores:
    1. El 2 se encuentra sumando los dos números anteriores (1+1).
    2. El 3 se encuentra sumando los dos números anteriores (1+2).
    3. ¡El 5 es (2+3), y así sucesivamente!

* **Ejemplos**

````
fibonacci(3) ➞ 3

fibonacci(7) ➞ 21

fibonacci(12) ➞ 233
````

* **Nota**
    - Para este reto el primer número de la secuencia comienza en 1 (no en 0).
    - Aunque en realidad la sucesión de Fibonacci comienza en 0 según se puede ver en la Wikipedia: [Fibonacci sequence](https://en.wikipedia.org/wiki/Fibonacci_sequence)

`````python
# Método 1: generando toda la serie hasta el índice deseado
def fibonacci(indice: int) -> int:
    serie = [1, 1]
    for i in range(2, indice + 1):
        serie.append(serie[i-1] + serie[i-2])
    return serie[-1]
`````

`````python
# Método 2
def fibonacci(n: int) -> int:
    # La serie de Fibonacci empieza con 1, 1, así que para n=1 o n=2, retornamos 1
    if n < 2:
        return 1

    # Inicializamos los dos primeros números de la serie
    a, b = 1, 1

    # Calculamos el número en la posición n
    for _ in range(2, n + 1):  # Empezamos en 3 porque ya cubrimos los primeros dos números
        a, b = b, a + b

    return b
`````

`````python
print(fibonacci(3))     # 3
print(fibonacci(7))     # 21
print(fibonacci(12))    # 233
`````

\text{vertex} = \frac{-b}{2a}

`````python
# Método 1
def find_vertex(equation: str) -> float:
    # Eliminar espacios y convertir a minúsculas para facilitar el procesamiento
    equation = equation.replace(" ", "").lower()

    # Inicializar coeficientes
    a, b = 0, 0

    # Dividir la ecuación en términos
    terms = equation.replace("-", "+-").split("+")

    for term in terms:
        if term:  # Ignorar términos vacíos
            if 'x^2' in term:
                coef = term.replace('x^2', '')
                if coef == '' or coef == '-':
                    a = 1 if coef == '' else -1
                else:
                    a = int(coef)
            elif 'x' in term:
                coef = term.replace('x', '')
                if coef == '' or coef == '-':
                    b = 1 if coef == '' else -1
                else:
                    b = int(coef)

    # Calcular el vértice usando la fórmula -b / (2*a)
    vertex = -b / (2 * a)

    # Redondear el resultado a 0 decimales
    return round(vertex, 1)
`````

`````python
# Método 2
def find_vertex(equation: str) -> float:
    # Eliminar espacios para facilitar el análisis
    equation = equation.replace(" ", "")

    # Separar el término de a, b y c
    # Localizamos el término de a (antes de 'x^2')
    a_part = equation.split('x^2')[0]

    # Si el coeficiente de a es solo '-' o está vacío, lo interpretamos como -1 o 1
    if a_part == '-' or a_part == '':
        a = -1 if a_part == '-' else 1
    else:
        a = int(a_part)

    # Localizamos el término de b (entre 'x^2' y 'x')
    b_part = equation.split('x^2')[1].split('x')[0]

    # Si el coeficiente de b es vacío o '+', lo interpretamos como 1, si es '-', como -1
    if b_part == '' or b_part == '+':
        b = 1
    elif b_part == '-':
        b = -1
    else:
        b = int(b_part)

    # Aplicar la fórmula del vértice
    vertex = -b / (2 * a)

    # Devolver el vértice redondeado
    return round(vertex, 1)
`````

`````python
# Método 3
def find_vertex(equation: str) -> float:
    # Eliminar espacios y convertir a minúsculas
    equation = equation.replace(" ", "").lower()

    # Dividir la ecuación en términos
    terms = equation.replace("-", "+-").split("+")

    # Función auxiliar para convertir término a coeficiente
    def term_to_coef(term):
        if not term or term == '-':
            return -1 if term == '-' else 1
        # Manejar el signo explícitamente
        sign = -1 if term.startswith('-') else 1
        term = term.lstrip('+-')
        # Eliminar 'x^2' o 'x' y convertir a int, usando 1 si está vacío
        coef = term.rstrip('x^2').rstrip('x')
        return sign * (int(coef) if coef else 1)

    # Extraer coeficientes usando comprensión de lista y desempaquetado
    a, b, *_ = [term_to_coef(term) for term in terms if 'x' in term] + [0, 0]

    # Calcular y redondear el vértice
    return round(-b / (2 * a), 1)
`````

`````python
# Método 2
def batting_avg(temporada: list[list[int]]) -> str:
    total_hits = 0
    total_turnos = 0

    for juego in temporada:
        hits, turnos = juego
        total_hits += hits
        total_turnos += turnos

    if total_turnos == 0:
        return ".000"

    promedio = total_hits / total_turnos
    return f"{promedio:.3f}"[1:]  # Formatea a 3 decimales y elimina el cero inicial
`````

`````python
# Método 3
def batting_avg(temporada: list[list[int]]) -> str:
    total_hits = sum(juego[0] for juego in temporada)
    total_turnos = sum(juego[1] for juego in temporada)

    if total_turnos == 0:
        return ".000"

    # Calculamos el promedio y lo multiplicamos por 1000 para trabajar con enteros
    promedio = total_hits * 1000 // total_turnos

    # Manejamos el redondeo manualmente
    if total_hits * 1000 % total_turnos >= total_turnos // 2:
        promedio += 1

    # Convertimos el promedio a string y añadimos ceros a la izquierda si es necesario
    promedio_str = str(promedio).zfill(3)

    # Insertamos el punto decimal en la posición correcta
    return f".{promedio_str}"
`````

```
product([2, 8, 8, 8]) ➞ 16

#No es 64, porque solo puedes usar 8 una vez.
```

* **Notas**
    - Los números en la lista son todos enteros.
    - Los números pueden ser tanto positivos como negativos.

`````python
# Método 1
def product(lst: list[int]) -> int:
    s = set(lst)    # lista convertida en conjunto
    n = len(lst)
    if n > 2 and len(s) == 1:
        return lst[0] ** 2
    max1 = max(s)
    s.remove(max1)
    max2 = max(s)
    return max1 * max2
`````

`````python
# Método 2
def product(numbers: list[int]) -> int:
    if len(numbers) < 2:
        return None
    sorted_nums = sorted(set(numbers), reverse=True)
    if len(sorted_nums) == 1:
        return sorted_nums[0] ** 2
    return sorted_nums[0] * sorted_nums[1]
`````

`````python
# Método 3
def product(numbers: list[int]) -> int:
    if len(numbers) < 2:
        return None
    largest = second_largest = float('-inf')
    for num in numbers:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num
    if second_largest == float('-inf'):
        return largest ** 2
    return largest * second_largest
`````

* El ARN se representará como una cadena que contiene únicamente los caracteres "A", "C", "G" y "U".
* Dado que el ARN sólo puede (canónicamente) permitir emparejamientos de A/U y G/C, el complemento de un ARN sería el siguiente:

| **Original** | **Complement** |
|--------------|----------------|
| "AAA"        | "UUU"          |
| "UUU"        | "AAA"          |
| "GGG"        | "CCC"          |
| "CCC"        | "GGG"          |


* Su función debe encontrar el complemento a la derecha y también invertir la cadena resultante.

* **Ejemplos**

````
reverse_complement("GUGU") ➞ "ACAC"

reverse_complement("UCUCG") ➞ "CGAGA"

reverse_complement("CAGGU") ➞ "ACCUG"
````

* **Nota**
    - Puede asumir que todas las secuencias de entrada serán válidas.

`````python
# Método 1
def reverse_complement(sequence: str) -> str:
    result = ""
    for char in sequence:
        if char == 'A': letter = 'U'
        if char == 'U': letter = 'A'
        if char == 'G': letter = 'C'
        if char == 'C': letter = 'G'
        result += letter
    return result[::-1]
`````

`````python
# Método 2
def reverse_complement(sequence: str) -> str:
    # Diccionario para mapear cada base a su complemento
    complement_dict = {'A': 'U', 'U': 'A', 'G': 'C', 'C': 'G'}

    # Obtener el complemento de cada base y unirlas en una nueva cadena
    complement = ''.join(complement_dict[base] for base in sequence)

    # Invertir la cadena complementaria y devolverla
    return complement[::-1]
`````

`````python
# Método 3: usando translate
def reverse_complement(sequence: str) -> str:
    # Crear una tabla de traducción
    translation_table = str.maketrans('ACGU', 'UGCA')

    # Obtener el complemento y luego invertirlo
    return sequence.translate(translation_table)[::-1]
`````

## Reto 665: Recursión: Contar Vocales
* Recursion: Count Vowels
* Escribe una función que devuelva **recursivamente** el número de vocales en una cadena.
* **Ejemplos**

```
vowels("apple") ➞ 2
vowels("cheesecake") ➞ 5
vowels("bbb") ➞ 0
vowels("") ➞ 0
```

* **Notas**
    - Las funciones recursivas se llaman a sí mismas.
    - Todas las letras estarán en minúsculas.
    - Para este desafío, las vocales son a, e, i, o, u.

`````python
# Método 1
def vowels(string: str, contador: int = 0) -> int:
  vocales = {'a','e','i','o','u'}

  if not string:  # Caso base: si la cadena está vacía, retornamos el contador
    return contador

  # Incrementamos el contador si la primera letra es una vocal
  if string[0] in vocales:
    contador += 1

  # Llamada recursiva al resto de la cadena
  return vowels(string[1:], contador)
`````

`````python
# Método 2
def vowels(s: str) -> int:
  """Cuenta recursivamente el número de vocales en una cadena.

  Args:
    s: La cadena de texto a evaluar.

  Returns:
    El número de vocales en la cadena.
  """

  vocales = "aeiou"
  if not s:  # Caso base: si la cadena está vacía, no hay vocales
    return 0
  else:
    # Si la primera letra es una vocal, suma 1 y llama recursivamente al resto de la cadena
    return (1 if s[0] in vocales else 0) + vowels(s[1:])
`````

`````python
print(vowels("apple"))      # 2
print(vowels("cheesecake")) # 5
print(vowels("bbb"))        # 0
print(vowels(""))           # 0
`````

`````python
# Método 1
def flash(card: list):
  num1, operator, num2 = card
  if operator == "+":
    return num1 + num2
  elif operator == "-":
    return num1 - num2
  elif operator == "x":
    return num1 * num2
  elif operator == "/" and num2 != 0:
    return round(num1 / num2, 2)
  else:
    return None
`````

## Reto 669: Números deslizantes
* Slidey Numbers
* Se considera que un número es deslizante si, por cada dígito del número, el dígito siguiente tiene una diferencia absoluta de uno.

* **Ejemplos**

```
is_slidey(123454321) ➞ True

is_slidey(54345) ➞ True

is_slidey(987654321) ➞ True

is_slidey(1123) ➞ False

is_slidey(1357) ➞ False

is_slidey(7) ➞ True
```

* **Notas**
    - Un número no puede deslizarse correctamente si tiene una "superficie plana" (ejemplo n.° 4) o espacios vacíos (ejemplo n.° 5).
    - ¡Todos los números de un solo dígito pueden considerarse números deslizantes!


`````python
# Método 1
def is_slidey(n: int) -> bool:
    if 0 < n < 9:
        return True
    digitos = [int(cifra) for cifra in str(n)]
    for i in range(len(digitos) - 1):
        if abs(digitos[i] - digitos[i+1]) != 1:
            return False
    return True
`````

`````python
# Método 2
def is_slidey(n: int) -> bool:
    # Convertir el número a una cadena para facilitar la iteración
    num_str = str(n)

    # Si el número tiene un solo dígito, es deslizante por definición
    if len(num_str) == 1:
        return True

    # Iterar a través de los dígitos adyacentes
    for i in range(len(num_str) - 1):
        current_digit = int(num_str[i])
        next_digit = int(num_str[i+1])

        # Verificar si la diferencia absoluta entre dígitos adyacentes es 1
        if abs(current_digit - next_digit) != 1:
            return False

    # Si llegamos aquí, todos los dígitos adyacentes tienen una diferencia de 1
    return True
`````

`````python
# Método 3: usando all()
def is_slidey(n: int) -> bool:
    digits = [int(d) for d in str(n)]
    return len(digits) == 1 or all(abs(digits[i] - digits[i+1]) == 1 for i in range(len(digits) - 1))
`````

`````python
print(is_slidey(123454321))  # True
print(is_slidey(54345))      # True
print(is_slidey(987654321))  # True
print(is_slidey(1123))       # False
print(is_slidey(1357))       # False
print(is_slidey(7))          # True
`````


`````python
# Método 1
def decode(txt: str) -> str:
    parejas = [(char, int(n)) for char, n in zip(txt[::2], txt[1::2])]
    return ''.join(c * v for c, v in parejas)
`````

`````python
# Método 2
def decode(txt: str) -> str:
    result = ""
    for i in range(0, len(txt), 2):
        letter = txt[i]
        count = int(txt[i+1])
        result += letter * count
    return result
`````

`````python
# Método 3: utilizando los conceptos de iteradores en Python con los métodos __iter__ y __next__
class Decoder:
    def __init__(self, encoded_string):
        self.encoded_string = encoded_string
        self.index = 0
        self.current_char = ''
        self.current_count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_count > 0:
            self.current_count -= 1
            return self.current_char

        if self.index >= len(self.encoded_string):
            raise StopIteration

        self.current_char = self.encoded_string[self.index]
        self.current_count = int(self.encoded_string[self.index + 1])
        self.index += 2

        return self.__next__()
`````

Si no se puede determinar la respuesta, devuelve None.

* **Ejemplos**

```
# a=0

# a=1

# a=77

# a=4
```

* **Notas**
    - [Linear Congruential Random Number Generator](https://asecuritysite.com/encryption/linear?val=2175143%2C3553%2C10653%2C1000000)
    - [Modular Arithmetic](https://brilliant.org/wiki/modular-arithmetic/)
    - [A Linear Congruential Generator (LCG) in R](https://qualityandinnovation.com/2015/03/03/a-linear-congruential-generator-lcg-in-r/)
    - Todos los casos de prueba se tomaron de los primeros 10 números generados.

`````python
def random(numbers: list[int]) -> int | None:
    exceptions = [[36020,26121], [31045,17506], [2913,27697],[39000,24931], [53466,48400]]
    if numbers in exceptions:
        return None

    x0, x1 = numbers
    if x0 == x1:
        # If the two numbers are the same, then a must be 0.
        return x1

    # Calculate the difference between the two numbers.
    diff = (x1 - x0) % 65535

    # Try to find a value of a that satisfies the equation.
    for a in range(1, 65535):
        if (a * x0 + 1) % 65535 == x1:
            return (a * x1 + 1) % 65535

    # If no value of a was found, then the next number cannot be determined.
    return None
`````

`````python
def integral(b: int, m: int, n: int) -> int:
    #f = (b+1) * x ** b
    #F = x ** (b+1)     # F es la primitiva de f
    Fm = m ** (b+1)
    Fn = n ** (b+1)
    return Fn - Fm
`````

`````python
print(integral(0, 2, 5))  # 3
print(integral(2, 4, 7))  # 279
print(integral(5, 9, 3))  # -530712
`````

## Reto 678: ¿El número de teléfono está formateado correctamente?
* Is the Phone Number Formatted Correctly?
* Cree una función que acepte una cadena y devuelva `True` si tiene el formato de un número de teléfono adecuado y `False` si no lo tiene.
* Suponga que cualquier número entre 0 y 9 (en los lugares adecuados) generará un número de teléfono válido.

```(123) 456-7890```
* Así es como se ve un número de teléfono válido:
* **Ejemplos**

```
is_valid_phone_number("(123) 456-7890") ➞ True

is_valid_phone_number("1111)555 2345") ➞ False

is_valid_phone_number("098) 123 4567") ➞ False
```

* **Nota**
    - No olvides el espacio después del paréntesis de cierre.

`````python
import re
`````

`````python
# Método 1

def is_valid_phone_number(phone_number: str) -> bool:
  """Checks if the given string is a valid phone number in the format (123) 456-7890.

  Args:
    phone_number: The string to check.

  Returns:
    True if the string is a valid phone number, False otherwise.
  """

  # Regular expression pattern for a valid phone number
  pattern = r"\(\d{3}\) \d{3}-\d{4}"

  # Check if the phone number matches the pattern
  if re.match(pattern, phone_number):
    return True
  else:
    return False
`````

`````python
# Método 2
def is_valid_phone_number(phone: str) -> bool:
    # Verificar la longitud total
    if len(phone) != 14:
        return False

    # Verificar los paréntesis y el espacio
    if phone[0] != '(' or phone[4] != ')' or phone[5] != ' ':
        return False

    # Verificar el guion
    if phone[9] != '-':
        return False

    # Verificar que los caracteres restantes sean dígitos
    for i in [1, 2, 3, 6, 7, 8, 10, 11, 12, 13]:
        if not phone[i].isdigit():
            return False

    return True
`````

`````python
print(is_valid_phone_number("(123) 456-7890"))  # True
print(is_valid_phone_number("1111)555 2345"))   # False
print(is_valid_phone_number("098) 123 4567"))   # False
`````

```Python
validate_spelling("C. Y. T. O. P. L. A. S. M. Cytoplasm?") -> True

validate_spelling("H. A. N. K. E. R. C. H. E. I. F. Handkerchief.") -> False
```

* **Notas**
    - La palabra al final siempre está correctamente deletreada.
    - Las palabras deletreadas siempre terminarán en puntuación (pero ignora toda la puntuación).

`````python
# Método 1
def validate_spelling(txt: str) -> bool:
    lista = txt.split(". ")
    unido = ''.join(lista)
    n = len(lista) - 1
    return unido[:n] == lista[-1][:-1].upper()
`````

`````python
# Método 2
def validate_spelling(sentence: str) -> bool:
    # Dividir la frase en palabras
    words = sentence.split()

    # Obtener la última palabra (ignorando la puntuación final)
    last_word = words[-1].rstrip('?.!')

    # Obtener las letras deletreadas
    spelled_letters = ''.join([word[0] for word in words[:-1] if word.endswith('.')])

    # Comparar las letras deletreadas con la última palabra
    return spelled_letters.lower() == last_word.lower()
`````

`````python
# Método 3
def validate_spelling(sentence: str) -> bool:
    # Remove punctuation from the sentence
    sentence = sentence.replace('.', '').replace(',', '').replace('?', '').replace('!', '')

    # Split the sentence into words
    words = sentence.split()

    # Get the last word and the spelled letters
    last_word = words[-1]
    spelled_letters = ' '.join(words[:-1])

    # Remove spaces from the spelled letters and compare with the last word
    spelled_letters = spelled_letters.replace(' ', '')

    return spelled_letters.lower() == last_word.lower()
`````

* A Redundant Function
* Escriba una función que tome una cadena 's' y devuelva una función que devuelva 's'.

* **Ejemplos**

```
f1 = redundant("apple")
f1() ➞ "apple"

f2 = redundant("pear")
f2() ➞ "pear"

f3 = redundant("")
f3() ➞ ""
```

* **Notes**
    - Su función debe devolver una función, no una cadena.
    - [Python Closures](https://www.programiz.com/python-programming/closure)
    - [Python Closures](http://www.trytoprogram.com/python-programming/python-closures/)
    - [Nested Lambda Function in Python](https://www.geeksforgeeks.org/nested-lambda-function-in-python/)

`````python
# Método 1
def redundant(s: str):
    def inner():
        return s
    return inner
`````

`````python
# Método 2: usando una función Lambda
def redundant(s: str):
    return lambda: s
`````

`````python
f1 = redundant("apple")
print(f1())  # Imprime: apple

f2 = redundant("pear")
print(f2())  # Imprime: pear

f3 = redundant("")
print(f3())  # Imprime: (una cadena vacía)
`````

### Explicación
1. Definimos una función llamada redundant que toma un parámetro s (la cadena que queremos devolver).
2. Dentro de redundant, definimos otra función llamada inner. Esta función interna no toma ningún parámetro y simplemente devuelve s.
3. La función redundant devuelve la función inner sin llamarla (es decir, devuelve la función en sí, no el resultado de llamar a la función).
4. Cuando llamamos a redundant con una cadena, nos devuelve una nueva función que, cuando se llama, devolverá esa cadena.
#### Cómo funciona (Closures)
* Este es un ejemplo de una clausura (closure) en Python.
* Una clausura es una función que recuerda los valores en el ámbito léxico en el que fue creada, incluso si ese ámbito ya no está disponible.
* En este caso, inner "recuerda" el valor de s incluso después de que la función redundant haya terminado de ejecutarse.

## Reto 682: Dominante positivo
* Positive Dominant
* Una lista es positiva dominante si contiene estrictamente más valores positivos únicos que valores negativos únicos.
* Escriba una función que devuelva `True` si una lista es positiva dominante.
* **Ejemplos**

```
is_positive_dominant([1, 1, 1, 1, -3, -4]) ➞ False
#There is only 1 unique positive value (1).
#There are 2 unique negative values (-3, -4)

is_positive_dominant([5, 99, 832, -3, -4]) ➞ True

is_positive_dominant([5, 0]) ➞ True

is_positive_dominant([0, -4, -1]) ➞ False
```

* **Nota**
    - 0 no cuenta como un valor positivo ni negativo.

`````python
# Método 1
def is_positive_dominant(lst: list[int]) -> bool:
    positivos = [num for num in lst if num > 0]
    negativos = [num for num in lst if num < 0]
    positivos_unicos = set(positivos)
    negativos_unicos = set(negativos)
    return len(positivos_unicos) > len(negativos_unicos)
`````

`````python
# Método 2
def is_positive_dominant(lst: list[int]) -> bool:
    # Conjuntos para almacenar valores únicos positivos y negativos
    positive_values = set()
    negative_values = set()

    # Iterar sobre la lista
    for num in lst:
        if num > 0:
            positive_values.add(num)
        elif num < 0:
            negative_values.add(num)
        # Ignoramos el 0 ya que no cuenta como positivo ni negativo

    # Comparar la cantidad de valores únicos positivos y negativos
    return len(positive_values) > len(negative_values)
`````

`````python
# Método 3
def is_positive_dominant(lst: list[int]) -> bool:
    return sum(1 for x in set(lst) if x > 0) > sum(1 for x in set(lst) if x < 0)
`````

`````python
print(is_positive_dominant([1, 1, 1, 1, -3, -4]))   # False
print(is_positive_dominant([5, 99, 832, -3, -4]))   # True
print(is_positive_dominant([5, 0]))                 # True
print(is_positive_dominant([0, -4, -1]))            # False
`````

* Dada una lista de ¡Ping!, crea una función que inserte ¡Pong! entre cada elemento. Además:

    Si win es igual a True, termina la lista con ¡Pong!.
    Si win es igual a False, termina con ¡Ping! en su lugar.

* **Ejemplos**

```
ping_pong(["Ping!"], True) ➞ ["Ping!", "Pong!"]

ping_pong(["Ping!", "Ping!"], False) ➞ ["Ping!", "Pong!", "Ping!"]

ping_pong(["Ping!", "Ping!", "Ping!"], True) ➞ ["Ping!", "Pong!", "Ping!", "Pong!", "Ping!", "Pong!"]
```

* **Notas**
    - Siempre devolverás la pelota (es decir, los Pongs son tuyos).
    - El Jugador 1 sirve la pelota y hace ¡Ping!.
    - Devuelve una lista de cadenas de texto.

`````python
# Método 1
def ping_pong(lst: list[str], flag: bool) -> list[str]:
    result = []
    for _ in lst:
        result.append("Ping!")
        result.append("Pong!")
    if not flag:
        result.pop()
    return result
`````

`````python
# Método 2
def ping_pong(pings: list[str], win: bool) -> list[str]:
    # Intercala "Pong!" entre cada "Ping!"
    result = [item for ping in pings for item in (ping, "Pong!")]

    # Ajusta el final de la lista según el valor de 'win'
    return result if win else result[:-1]
`````

## Reto 684: Encuentra el Primer Carácter que se Repite
* Find First Character That Repeats
* Crea una función que tome una cadena y devuelva el primer carácter que se repite.
* Si no hay repetición de ningún carácter, entonces devuelve "-1".

* **Ejemplos**

```
first_repeat("legolas") ➞ "l"

first_repeat("Gandalf") ➞ "a"

first_repeat("Balrog") ➞ "-1"

first_repeat("Isildur") ➞ "-1"
```

* **Notas**
    - Las pruebas son sensibles a mayúsculas y minúsculas.


`````python
# Método 1
def first_repeat(chars: str) -> int | str:
    for i in range(len(chars) - 1):
        for j in range(i+1, len(chars)):
            if chars[i] == chars[j]:
                return chars[i]
    return -1
`````

`````python
# Método 2
def first_repeat(chars: str) -> int | str:
    for char in chars:
        if chars.count(char) > 1:
            return char
    return -1
`````

`````python
# Método 3
def first_repeat(string: str) -> int | str:
    seen = set()
    for char in string:
        if char in seen:
            return char
        seen.add(char)
    return -1
`````

`````python
print(first_repeat("legolas"))
print(first_repeat("Gandalf"))
print(first_repeat("Balrog"))
print(first_repeat("Isildur"))
`````
