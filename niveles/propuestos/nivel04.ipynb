## Resuelto 301: Eliminar duplicados de una lista de invitados
* Cree una función que tome una lista de nombres de invitados y elimine todos los elementos duplicados.
* La lista que se retorna ha de tener el mismo orden secuencial que la lista original (excepto duplicados).
* Se ha de distinguir entre mayúsculas y minúsculas.
* Ejemplos:
    1. invitados_unicos([1, 0, 1, 0]) ➞ [1, 0]
    2. invitados_unicos(['Ana', 'Luis', 'Bea']) ➞ ['Ana', 'Luis', 'Bea']
    3. invitados_unicos(['Roma', 'Paris', 'Roma']) ➞ ['Roma', 'Paris']
    4. invitados_unicos([True, 1, False, 0, 'true', 'True']) ➞ [True, False, 'true', 'True']

````python
# Método 1: con dict.fromkeys() para mantener orden
def invitados_unicos(lista):
    return list(dict.fromkeys(lista))
````

````python
# Método 2: con un bucle y un set de vistos
def invitados_unicos(lista):
    vistos = set()
    resultado = []
    for x in lista:
        if x not in vistos:
            vistos.add(x)
            resultado.append(x)
    return resultado
````

````python
print(invitados_unicos([1, 0, 1, 0]))
print(invitados_unicos(['Ana', 'Luis', 'Bea']))
print(invitados_unicos(['Roma', 'Paris', 'Roma']))
print(invitados_unicos([True, 1, False, 0, 'true', 'True']))
````

### Comentario
- `dict.fromkeys(lista)` crea un diccionario usando los elementos de `lista` como claves (los valores no importan aquí). Como un diccionario no puede tener claves repetidas, esto elimina automáticamente los duplicados; y como desde Python 3.7 los diccionarios recuerdan el orden en que se insertaron sus claves, al convertir el resultado de nuevo a lista con `list(...)` se conserva el orden original. Es una forma muy compacta de "eliminar duplicados manteniendo el orden", alternativa al patrón con un `set` de "vistos" que ya vimos en el Nivel 3 (Resuelto 197).

## Resuelto 302: Distancia entre dos drones
* En este desafío, debes encontrar la distancia entre dos drones que vuelan en un plano cartesiano.
* Conociendo las coordenadas de ambos, hay que aplicar el teorema de Pitágoras para encontrar la distancia entre ellos.
* Dado que dos diccionarios `a` y `b` son las coordenadas de los dos drones (`x` e `y`), implemente una función que devuelva la distancia entre ellos, redondeada a la milésima más cercana.
* Ejemplos:
    1. distancia_drones({'x': -2, 'y': 1}, {'x': 4, 'y': 3}) ➞ 6.325
    2. distancia_drones({'x': 0, 'y': 0}, {'x': 1, 'y': 1}) ➞ 1.414
    3. distancia_drones({'x': 10, 'y': -5}, {'x': 8, 'y': 16}) ➞ 21.095
    4. distancia_drones({'x': 4, 'y': 0}, {'x': 0, 'y': 3}) ➞ 5.0

````python
# Método 1: aplicando el teorema de Pitágoras
import math

def distancia_drones(a, b):
    return round(math.sqrt((a["x"] - b["x"]) ** 2 + (a["y"] - b["y"]) ** 2), 3)
````

````python
# Método 2: con ** 0.5 en vez de math.sqrt
def distancia_drones(a, b):
    suma_cuadrados = (a["x"] - b["x"]) ** 2 + (a["y"] - b["y"]) ** 2
    return round(suma_cuadrados ** 0.5, 3)
````

````python
print(distancia_drones({'x': -2, 'y': 1}, {'x': 4, 'y': 3}))
print(distancia_drones({'x': 0, 'y': 0}, {'x': 1, 'y': 1}))
print(distancia_drones({'x': 10, 'y': -5}, {'x': 8, 'y': 16}))
print(distancia_drones({'x': 4, 'y': 0}, {'x': 0, 'y': 3}))
````

## Resuelto 303: Cifras pares vs. impares de un ticket
* Dado el número de un ticket, devuelva "impar" si la suma de todos los dígitos impares es mayor que la suma de todos los dígitos pares.
* Devuelva "par" si la suma de los dígitos pares es mayor que la suma de los dígitos impares.
* Devuelva "igual" si ambas sumas son iguales.
* Ejemplos:
    1. par_impar_ticket(97428) ➞ "impar"
    2. par_impar_ticket(81961) ➞ "par"
    3. par_impar_ticket(54870) ➞ "igual"

````python
# Método 1: sumando dígitos pares e impares por separado
def par_impar_ticket(numero):
    pares = sum(int(d) for d in str(numero) if int(d) % 2 == 0)
    impares = sum(int(d) for d in str(numero) if int(d) % 2 != 0)
    if impares > pares:
        return "impar"
    if pares > impares:
        return "par"
    return "igual"
````

````python
# Método 2: con map() para convertir dígitos
def par_impar_ticket(numero):
    digitos = list(map(int, str(numero)))
    pares = sum(d for d in digitos if d % 2 == 0)
    impares = sum(d for d in digitos if d % 2 != 0)
    if impares > pares:
        return "impar"
    if pares > impares:
        return "par"
    return "igual"
````

````python
print(par_impar_ticket(97428))
print(par_impar_ticket(81961))
print(par_impar_ticket(54870))
````

## Resuelto 304: Aplicar comisiones
* Cree una función que reciba una lista de importes de venta y un porcentaje de comisión `c`.
* La función devolverá una lista con el importe de las comisiones correspondientes.
* No habrá números con muchos decimales incómodos con los que lidiar, solo 0,5.
* Ejemplos:
    1. obtener_comision([2, 4, 6, 11], "50%") ➞ [1.0, 2.0, 3.0, 5.5]
    2. obtener_comision([6, 10, 20, 40], "25%") ➞ [1.5, 2.5, 5.0, 10.0]
    3. obtener_comision([100], "45%") ➞ [45.0]

````python
# Método 1: convirtiendo el porcentaje y aplicándolo
def obtener_comision(nums, c):
    pct = float(c.strip("%")) / 100
    return [n * pct for n in nums]
````

## Resuelto 306: Comparar por su código
* Cree una función que compare dos nombres de usuario en función de la suma de sus códigos ASCII y devuelva el nombre con la suma ASCII más pequeña.
* Las dos cadenas proporcionadas como argumentos tendrán sumas ASCII estrictamente diferentes.
* Ejemplos:
    1. menor_ascii_usuario(['indemne', 'abadejo']) ➞ "abadejo"
    2. menor_ascii_usuario(['desidia', 'diletante']) ➞ "desidia"
    3. menor_ascii_usuario(['fatuo', 'falaz']) ➞ "falaz"

````python
# Método 1: comparando sumas ASCII con min()
def menor_ascii_usuario(lista):
    return min(lista, key=lambda s: sum(ord(c) for c in s))
````

````python
# Método 2: con una función auxiliar que suma ASCII
def suma_ascii(palabra):
    total = 0
    for c in palabra:
        total += ord(c)
    return total

def menor_ascii_usuario(lista):
    if suma_ascii(lista[0]) < suma_ascii(lista[1]):
        return lista[0]
    return lista[1]
````

````python
print(menor_ascii_usuario(['indemne', 'abadejo']))
print(menor_ascii_usuario(['desidia', 'diletante']))
print(menor_ascii_usuario(['fatuo', 'falaz']))
````

### Comentario
- `min(lista, key=lambda s: sum(ord(c) for c in s))` combina dos ideas ya conocidas: el parámetro `key` de `sorted()`/`min()`/`max()` (Nivel 3, Resuelto 255) y las funciones lambda (Nivel 2, Resuelto 124). Aquí la función lambda no se guarda en una variable, sino que se escribe directamente donde se necesita: para cada cadena de la lista, calcula la suma de sus códigos ASCII, y `min()` devuelve la cadena cuya suma sea más pequeña.

## Resuelto 307: Ordenar dígitos de un código en orden ascendente
* Cree una función que tome cualquier número no negativo como argumento y lo devuelva con sus dígitos en orden ascendente.
* El orden ascendente es cuando se ordena de menor a mayor.
* Ejemplos:
    1. orden_ascendente_digitos(123) ➞ 123
    2. orden_ascendente_digitos(1254859723) ➞ 1223455789
    3. orden_ascendente_digitos(73065) ➞ 3567

````python
# Método 1: convirtiendo a str, ordenando y volviendo a int
def orden_ascendente_digitos(numero):
    return int("".join(sorted(str(numero))))
````

````python
# Método 2: sin convertir a cadena, extrayendo dígitos
def orden_ascendente_digitos(numero):
    if numero == 0:
        return 0
    digitos = []
    while numero > 0:
        digitos.append(numero % 10)
        numero //= 10
    digitos.sort()
    resultado = 0
    for d in digitos:
        resultado = resultado * 10 + d
    return resultado
````

````python
print(orden_ascendente_digitos(123))
print(orden_ascendente_digitos(1254859723))
print(orden_ascendente_digitos(73065))
````

## Resuelto 308: La secuencia más larga de Unos consecutivos
* Escriba una función que devuelva la secuencia más larga de unos consecutivos en una cadena binaria.
* Ejemplos:
    1. unos("01100001011000") ➞ "11"
    2. unos("100100100") ➞ "1"
    3. unos("00000") ➞ ""

````python
# Método 1: con split() y max()
def unos(cadena):
    return max(cadena.split("0"), key=len)
````

````python
# Método 2: recorriendo la cadena
def unos(cadena):
    max_unos = ""
    actual = ""
    for c in cadena:
        if c == "1":
            actual += c
        else:
            if len(actual) > len(max_unos):
                max_unos = actual
            actual = ""
    if len(actual) > len(max_unos):
        max_unos = actual
    return max_unos
````

````python
# Método 3: recorriendo la cadena contando rachas
def unos(cadena):
    contador = 0
    maximo = 0
    for d in cadena:
        if d == "1":
            contador += 1
            if contador > maximo:
                maximo = contador
        else:
            contador = 0
    return "1" * maximo
````

````python
print(unos("01100001011000"))
print(unos("100100100"))
print(unos("00000"))
````

### Comentario
- El Método 1 usa un truco ingenioso: `cadena.split("0")` corta la cadena binaria en trozos, usando cada "0" como separador. Lo que queda entre dos "0" son precisamente las rachas de "1" consecutivos. Como `max(..., key=len)` ya vimos en el Resuelto 255, basta con pedirle el trozo más largo.
- Los Métodos 2 y 3 resuelven lo mismo "a mano", recorriendo la cadena carácter a carácter y llevando la cuenta de la racha actual, reiniciándola cada vez que aparece un "0".

## Resuelto 310: Enmascarar el IBAN
* Su tarea es crear una función que tome una cadena (un IBAN), transforme todos los caracteres excepto los últimos cuatro en "#" y devuelva la nueva cadena enmascarada.
* La función debe aceptar una cadena de cualquier longitud.
* Una cadena vacía debería devolver una cadena vacía.
* Ejemplos:
    1. enmascarar_iban("ES7620770024003102575766") ➞ "####################5766"
    2. enmascarar_iban("64607935616") ➞ "#######5616"
    3. enmascarar_iban("1") ➞ "1"
    4. enmascarar_iban("") ➞ ""

````python
# Método 1: reemplazando todo menos los últimos 4 con #
def enmascarar_iban(cadena):
    if len(cadena) <= 4:
        return cadena
    return "#" * (len(cadena) - 4) + cadena[-4:]
````

````python
print(enmascarar_iban("ES7620770024003102575766"))
print(enmascarar_iban("64607935616"))
print(enmascarar_iban("1"))
print(enmascarar_iban(""))
````

## Resuelto 312: Sin ingredientes en común
* Escriba una función que devuelva `True` si dos listas de la compra no tienen ningún ingrediente en común.
* De lo contrario, devuelve `False`.
* Ejemplos:
    1. no_comparten_ingredientes(['arroz', 'pollo', 'sal'], ['pasta', 'atún', 'aceite']) ➞ True
    2. no_comparten_ingredientes(['arroz', 'pollo', 'sal'], ['sal', 'pimienta', 'ajo']) ➞ False
    3. no_comparten_ingredientes(['huevos', 'leche'], ['pan', 'mantequilla']) ➞ True

````python
# Método 1: con set y isdisjoint()
def no_comparten_ingredientes(lst1, lst2):
    return set(lst1).isdisjoint(set(lst2))
````

````python
print(no_comparten_ingredientes(['arroz', 'pollo', 'sal'], ['pasta', 'atún', 'aceite']))
print(no_comparten_ingredientes(['arroz', 'pollo', 'sal'], ['sal', 'pimienta', 'ajo']))
print(no_comparten_ingredientes(['huevos', 'leche'], ['pan', 'mantequilla']))
````

### Comentario
- `set(lst1).isdisjoint(set(lst2))` comprueba si dos conjuntos **no tienen ningún elemento en común**, devolviendo `True` en ese caso. Es el complementario de `issubset()` (Nivel 3, Resuelto 242): mientras `issubset()` pregunta "¿todos los elementos de uno están en el otro?", `isdisjoint()` pregunta "¿no comparten ninguno?".

## Resuelto 313: Convertidor de Velocidad entre Vehículos
* En este desafío, tienes que convertir una velocidad medida con un vehículo a la velocidad equivalente que alcanzaría otro vehículo con el mismo esfuerzo relativo.
* Para realizar la conversión, debes dividir la velocidad por el factor del vehículo en el que se ha medido y luego multiplicar el resultado por el factor del otro vehículo.
* Consulte la tabla para ver los factores:
| Vehículo   | factor |
|------------|--------|
| Bicicleta  | 1.0    |
| Patinete   | 1.4    |
| Moto       | 6.0    |
| Coche      | 8.5    |
| Tren       | 25.0   |
* Dada una velocidad medida en vehiculo_a, devuelve el valor convertido para vehiculo_b redondeado a la centésima más cercana.
* Ejemplos:
    1. convertir_velocidad("Bicicleta", 20, "Moto") ➞ 120.0
    2. convertir_velocidad("Coche", 90, "Tren") ➞ 264.71
    3. convertir_velocidad("Bicicleta", 15, "Patinete") ➞ 21.0
    4. convertir_velocidad("Moto", 60, "Bicicleta") ➞ 10.0

````python
# Método 1: con un diccionario de factores
def convertir_velocidad(vehiculo_a, velocidad, vehiculo_b):
    factores = {"Bicicleta": 1.0, "Patinete": 1.4, "Moto": 6.0, "Coche": 8.5, "Tren": 25.0}
    return round(velocidad / factores[vehiculo_a] * factores[vehiculo_b], 2)
````

## Resuelto 315: Carrera de Tortugas
* Iker y Nadia tienen tortugas de carreras.
* Cada uno tiene tres: una lenta `s`, una mediana `m` y una rápida `f`.
* Aunque las tortugas de Iker son un poco más fuertes que las de Nadia, Nadia tiene un truco bajo la manga. Su plan es el siguiente:
1. Ronda 1: [s, f] Sacrificar su tortuga más lenta contra la más rápida de Iker.
2. Ronda 2: [m, s] Usar su tortuga intermedia contra la más lenta de Iker.
3. Ronda 3: [f, m] Usar su tortuga más rápida contra la intermedia de Iker.
* Crea una función que determine si el plan de Nadia funcionará devolviendo `True` si Nadia gana 2 de 3 juegos.
* Los inputs de la función son: [s, m, f] para Nadia y [s, m, f] para Iker.
* Nadia gana si la velocidad de su tortuga competidora excede estrictamente la velocidad de la tortuga de Iker.
* Iker siempre jugará en este orden: [f, s, m].
* Ejemplos:
    1. nadia_gana([3, 5, 10], [4, 7, 11]) ➞ True
    2. nadia_gana([6, 8, 9], [7, 12, 14]) ➞ False
    3. nadia_gana([1, 8, 20], [2, 9, 100]) ➞ True

````python
# Método 1: simulando las 3 rondas
def nadia_gana(nadia, iker):
    victorias = 0
    # Ronda 1: lenta de Nadia vs rápida de Iker
    if nadia[0] > iker[2]:
        victorias += 1
    # Ronda 2: mediana de Nadia vs lenta de Iker
    if nadia[1] > iker[0]:
        victorias += 1
    # Ronda 3: rápida de Nadia vs mediana de Iker
    if nadia[2] > iker[1]:
        victorias += 1
    return victorias >= 2
````

````python
print(nadia_gana([3, 5, 10], [4, 7, 11]))
print(nadia_gana([6, 8, 9], [7, 12, 14]))
print(nadia_gana([1, 8, 20], [2, 9, 100]))
````

## Resuelto 316: Encuentra la Cantidad de Dígitos de un Código de Barras
* Cree una función que devuelva un número entero correspondiente a la cantidad de dígitos en el código de barras (un número entero) dado.
* Intente resolver este desafío sin usar cadenas de texto.
* Ejemplos:
    1. cantidad_digitos_barras(1000) ➞ 4
    2. cantidad_digitos_barras(12) ➞ 2
    3. cantidad_digitos_barras(1305981031) ➞ 10
    4. cantidad_digitos_barras(0) ➞ 1

````python
# Método 1: sin usar str: con división sucesiva
def cantidad_digitos_barras(num):
    if num == 0:
        return 1
    contador = 0
    while num > 0:
        num //= 10
        contador += 1
    return contador
````

````python
# Método 2: convirtiendo a cadena
def cantidad_digitos_barras(num):
    return len(str(num))
````

````python
print(cantidad_digitos_barras(1000))
print(cantidad_digitos_barras(12))
print(cantidad_digitos_barras(1305981031))
print(cantidad_digitos_barras(0))
````

````python
# Método 1: con translate() y str.maketrans()
def leet_speak(cadena):
    tabla = str.maketrans({"e": "3", "l": "1", "t": "7", "o": "0", "a": "4", "E": "3", "L": "1", "T": "7", "O": "0", "A": "4"})
    return cadena.translate(tabla)
````

````python
# Método 2: con replace() encadenado
def leet_speak(cadena):
    cambios = [("e","3"),("l","1"),("t","7"),("o","0"),("a","4"),("E","3"),("L","1"),("T","7"),("O","0"),("A","4")]
    for viejo, nuevo in cambios:
        cadena = cadena.replace(viejo, nuevo)
    return cadena
````

- `str.maketrans(diccionario)` crea una "tabla de traducción": un diccionario especial que indica qué carácter debe sustituir a cada otro. `cadena.translate(tabla)` aplica esa tabla de una sola vez sobre toda la cadena.

## Resuelto 332: Contar las Antenas
* Cree una función que cuente el número de antenas en una matriz de caracteres.
* Se te proporciona una matriz 2D.
* Las antenas tienen una longitud de dos caracteres.
* Las antenas están compuestas únicamente del carácter |.
* Si en algún caso no hay antena, retorna 0.
* Ejemplos:
    1. contar_antenas([['     ||         '], ['||   ||        ||'], ['||   ||   ||   ||'], ['||   ||   ||   ||']]) ➞ 4
    2. contar_antenas([['||'], ['||']]) ➞ 1
    3. contar_antenas([[], []]) ➞ 0

````python
# Método 1: contando posiciones de columna únicas (no apariciones totales)
def contar_antenas(matriz):
    posiciones = set()
    for fila in matriz:
        for celda in fila:
            j = 0
            while True:
                idx = celda.find("||", j)
                if idx == -1:
                    break
                posiciones.add(idx)
                j = idx + 2
    return len(posiciones)
````

````python
print(contar_antenas([['     ||         '], ['||   ||        ||'], ['||   ||   ||   ||'], ['||   ||   ||   ||']]))
print(contar_antenas([['||'], ['||']]))
print(contar_antenas([[], []]))
````

### Comentario
- `celda.find("||", j)` acepta un segundo argumento: la posición a partir de la cual empezar a buscar. Repetir la búsqueda actualizando `j = idx + 2` en cada vuelta permite encontrar **todas** las apariciones de `"||"` en la cadena, no solo la primera, sin necesidad de expresiones regulares.

## Resuelto 345: Prevenir el Colapso de la Calculadora
* Dividir por 0 es un error enorme y debe evitarse a toda costa.
* Cree una función que reciba una expresión matemática en forma de cadena.

* La función previene un posible error de división por cero y en ese caso devuelve la frase "División por cero".

````python
# Método 1: con eval() y try/except
def calculadora_segura(expresion):
    try:
        resultado = eval(expresion)
        return resultado
    except ZeroDivisionError:
        return "División por cero"
````

### Comentario
- `eval(expresion)` toma una cadena de texto y la **ejecuta como si fuera código Python real**, devolviendo su resultado; por eso `eval("3 * 8 - 4")` calcula literalmente esa operación.
- ⚠️ **Importante**: `eval()` ejecutará *cualquier* código Python válido que reciba, no solo operaciones aritméticas. Úsala solo con expresiones de confianza; **nunca** con texto introducido directamente por un usuario no confiable, ya que podría ejecutar código arbitrario (incluso borrar archivos o acceder a datos sensibles). En una aplicación real, para evaluar expresiones matemáticas de usuarios se usan alternativas más seguras, como escribir un pequeño analizador propio o bibliotecas especializadas.
- Fíjate también en que el `except` captura específicamente `ZeroDivisionError`, y no cualquier error: así, si `eval()` fallara por otro motivo distinto (por ejemplo, una expresión mal escrita), el programa seguiría avisando de ese otro error en lugar de ocultarlo silenciosamente.

    4. primer_ingrediente_unico("") ➞ False

````python
# Método 1: con count()
def primer_ingrediente_unico(cadena):
    if not cadena:
        return False
    lista = cadena.split()
    for x in lista:
        if lista.count(x) == 1:
            return x
    return False
````

````python
# Método 1: comparando sumas de izquierda y derecha
def mayor_peso_sube_baja(lista):
    mitad = len(lista) // 2
    izq = sum(x for x in lista[:mitad] if isinstance(x, int))
    der = sum(x for x in lista[mitad+1:] if isinstance(x, int))
    if izq > der: return "izquierda"
    if der > izq: return "derecha"
    return "equilibrado"
````

## Resuelto 361: Comprueba si un Número es Perfecto
* Cree una función que devuelva `True` si un número es un número perfecto y `False` en caso contrario.
* Un número perfecto es un entero positivo que es igual a la suma de sus divisores propios (todos los divisores excepto él mismo).
* Los primeros números perfectos son: 6, 28, 496, 8128.
* Ejemplos:
    1. es_perfecto(18) ➞ False
    2. es_perfecto(28) ➞ True
    3. es_perfecto(6) ➞ True
    4. es_perfecto(496) ➞ True
    5. es_perfecto(4) ➞ False
    6. es_perfecto(1) ➞ False

````python
# Método 1: sumando divisores propios
def es_perfecto(n):
    if n < 2:
        return False
    suma = 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            suma += i
            if i != n // i:
                suma += n // i
    return suma == n
````

````python
print(es_perfecto(18))
print(es_perfecto(28))
print(es_perfecto(6))
print(es_perfecto(496))
print(es_perfecto(4))
print(es_perfecto(1))
````

### Comentario
- En vez de comprobar todos los números desde `2` hasta `n - 1` para encontrar los divisores (lo cual sería muy lento para números grandes), el bucle solo llega hasta `int(n ** 0.5) + 1`: si `n` tiene un divisor `i` mayor que su raíz cuadrada, el divisor "pareja" (`n // i`) será necesariamente menor que la raíz cuadrada, así que ya se habría encontrado antes. Por cada divisor `i` que se encuentra, se suman tanto `i` como su pareja `n // i` (salvo que coincidan, como en los cuadrados perfectos del Resuelto 341, para no contarlo dos veces).
- Esta optimización (buscar divisores solo hasta la raíz cuadrada) es una técnica clásica que hace que el cálculo sea mucho más rápido cuando `n` es un número grande.

## Resuelto 375: Recursión: Suma de División
* Dado un número, devuelve la suma total de ese número dividido entre cada número entero entre 1 y 10.
* No utilices la función integrada sum().
* Use una función recursiva para resolverlo.
* Ejemplos:
    1. div_sum(100) ➞ 292.89682539682536
    2. div_sum(200) ➞ 585.7936507936507
    3. div_sum(10) ➞ 29.289682539682538

````python
# Método 1: recursivamente: n/i + div_sum(n, i-1)
def div_sum(n, i=10):
    if i == 0:
        return 0
    return n / i + div_sum(n, i - 1)
````

````python
print(div_sum(100))
print(div_sum(200))
print(div_sum(10))
````

### Comentario
- Esta función recursiva usa un segundo parámetro, `i=10`, con un **valor por defecto**, que actúa como contador interno: en la primera llamada (`div_sum(100)`) no hace falta indicarlo y empieza en 10, y en cada llamada recursiva se le resta 1 (`div_sum(n, i - 1)`), hasta llegar al caso base (`i == 0`).
- Es una variante del patrón recursivo ya conocido (Nivel 2, Resueltos 173, 199 y 200): en vez de reducir directamente el argumento principal (`n`), se reduce un segundo parámetro que sirve de contador, dejando `n` sin cambios en todas las llamadas.

## Resuelto 385: Números Felices
* Un número es feliz cuando, al sumar el cuadrado de cada uno de sus dígitos repetidamente, se llega finalmente a 1.
* Por ejemplo: 19 → 1²+9² = 82 → 8²+2² = 68 → 6²+8² = 100 → 1²+0²+0² = 1 → ¡es feliz!
* Dado un entero positivo n, implemente una función que devuelva `True` si el número es feliz y `False` si no lo es (si entra en un bucle que nunca llega a 1).
* Ejemplos:
    1. es_feliz(19) ➞ True
    2. es_feliz(4) ➞ False
    3. es_feliz(7) ➞ True
    4. es_feliz(2) ➞ False

````python
# Método 1: iterando hasta llegar a 1 o detectar bucle
def es_feliz(n):
    vistos = set()
    while n != 1:
        if n in vistos:
            return False
        vistos.add(n)
        n = sum(int(d) ** 2 for d in str(n))
    return True
````

````python
print(es_feliz(19))
print(es_feliz(4))
print(es_feliz(7))
print(es_feliz(2))
````

### Comentario
- El bucle `while n != 1:` podría no terminar nunca si el número no es feliz, ya que entraría en un ciclo que se repite para siempre sin llegar a 1. Por eso se usa un conjunto `vistos` para llevar la cuenta de los números por los que ya se ha pasado: si se vuelve a llegar a un número ya visto, es señal de que se ha entrado en un ciclo, y se puede devolver `False` con seguridad, sin quedarse esperando indefinidamente.
- Esta técnica de "guardar lo ya visto para detectar ciclos" es un patrón general muy útil en cualquier bucle cuyo final no esté garantizado de antemano.

## Resuelto 386: Validación de Suma de Comprobación
* Un dígito de control se utiliza como un checksum muy simple para asegurar que un código no se corrompa al transmitirse.
* Así es como funciona:
1. Se suman todos los dígitos del código (sin el último).
2. Si la suma es múltiplo de 10, el dígito de control es 0; si no, es 10 menos el resto de dividir la suma entre 10.
* El dígito de control se añade al final del código.
* Crea una función que valide si un código es válido, utilizando su dígito de control.
* Todos los inputs tendrán una longitud de 5 caracteres.
* Ejemplos:

    2. validar_codigo("12322") ➞ True
    3. validar_codigo("99990") ➞ False
    4. validar_codigo("99994") ➞ True

````python
# Método 1: calculando el dígito de control
def validar_codigo(codigo):
    digitos = [int(c) for c in codigo]
    suma = sum(digitos[:-1])
    check = 0 if suma % 10 == 0 else 10 - (suma % 10)
    return digitos[-1] == check
````

````python
# Método 1: con isinstance()
import datetime

def tipo_de_fecha(variable):
    if isinstance(variable, datetime.datetime):
        return "datetime"
    if isinstance(variable, datetime.date):
        return "date"
    if isinstance(variable, str):
        return "string"
    if isinstance(variable, int):
        return "integer"
````

````python
print(tipo_de_fecha(5))
print(tipo_de_fecha("Hola"))
print(tipo_de_fecha(datetime.date(2018, 1, 1)))
print(tipo_de_fecha(datetime.datetime(2018, 1, 1, 10, 30)))
````

### Comentario
- El orden de las comprobaciones con `isinstance()` es importante aquí: en el módulo `datetime`, el tipo `datetime.datetime` es en realidad una **subclase** de `datetime.date` (todo `datetime` "es también" un `date`, igual que vimos que `bool` es un subtipo de `int` en el Nivel 1, Resuelto 36). Por eso el código comprueba primero `isinstance(variable, datetime.datetime)`: si se comprobara antes `datetime.date`, un objeto `datetime.datetime` pasaría esa comprobación igualmente (por ser también un `date`) y el resultado sería incorrecto.

    3. completar_hex("110010100010") ➞ "110010100010"

````python
# Método 1: rellenando con ceros a la izquierda
def completar_hex(cadena):
    resto = len(cadena) % 4
    if resto == 0:
        return cadena
    return "0" * (4 - resto) + cadena
````

## Resuelto 395: Catástrofe demográfica escolar
* Un colegio empieza el curso con 100 plazas disponibles y 100 alumnos matriculados.
* Cada año las plazas aumentan en una cantidad fija, pero los alumnos matriculados crecen por un porcentaje.
* Tu tarea es descubrir cuándo las plazas dejarán de ser suficientes.
* La función pasará:
- plazas_extra: un número entero: aumento de plazas por año.
- crecimiento: un número de punto flotante: multiplicador de crecimiento de alumnos por año.
* Devuelve el año en que ocurre el desbordamiento.
* Si el número de alumnos y las plazas son iguales, eso no es una catástrofe.
* Ejemplos:
    1. colapso_escolar(4255, 1.41) ➞ 20
    2. colapso_escolar(9433, 1.09) ➞ 108
    3. colapso_escolar(5879, 1.77) ➞ 12

````python
# Método 1: simulando año a año
def colapso_escolar(plazas_extra, crecimiento):
    plazas = 100
    alumnos = 100
    anio = 0
    while alumnos <= plazas:
        plazas += plazas_extra
        alumnos = int(alumnos * crecimiento)
        anio += 1
    return anio
````

````python
print(colapso_escolar(4255, 1.41))
print(colapso_escolar(9433, 1.09))
print(colapso_escolar(5879, 1.77))
````

* Crea una función que divida una cadena en partes de tamaño n.
* Para entradas que no se dividen de manera uniforme en partes de tamaño N, el último elemento del array tendrá una cadena "restante" (más corta).
* Ejemplos:
    1. particionar("Roma", 2) ➞ ['Ro', 'ma']
    2. particionar("Victoria", 4) ➞ ['Vict', 'oria']
    3. particionar("esternocleidomastoideo", 7) ➞ ['esterno', 'cleidom', 'astoide', 'o']
    4. particionar("Rio", 2) ➞ ['Ri', 'o']

````python
# Método 1: con range() y slicing
def particionar(cadena, n):
    return [cadena[i:i+n] for i in range(0, len(cadena), n)]
````

````python
print(particionar("Roma", 2))
print(particionar("Victoria", 4))
print(particionar("esternocleidomastoideo", 7))
print(particionar("Rio", 2))
````

### Comentario
- `[cadena[i:i+n] for i in range(0, len(cadena), n)]` es un patrón muy reutilizable para **dividir cualquier secuencia en trozos de tamaño fijo**: `range(0, len(cadena), n)` genera las posiciones de inicio de cada trozo (0, n, 2n, ...), y el slicing `cadena[i:i+n]` extrae cada trozo; si la cadena no es múltiplo exacto de `n`, el último trozo, más corto, se obtiene automáticamente sin ningún caso especial, porque el slicing nunca da error aunque se pida más allá del final (como ya vimos en el Nivel 2, Resuelto 147).

## Resuelto 398: ¿Existe un precio?
* Crea una función que valide si un precio dado existe y podría representar una cantidad de la vida real.
* Las entradas se darán como una cadena.
* Aceptar números como .5 y 0003.
* Ejemplos:
    1. valid_str_precio("3.2") ➞ True
    2. valid_str_precio("324") ➞ True
    3. valid_str_precio(".5") ➞ True
    4. valid_str_precio("0003") ➞ True
    5. valid_str_precio("54..4") ➞ False
    6. valid_str_precio("precio") ➞ False

````python
# Método 1: con try/except y float()
def valid_str_precio(cadena):
    try:
        float(cadena)
        return cadena.count(".") <= 1
    except ValueError:
        return False
````

````python
print(valid_str_precio("3.2"))
print(valid_str_precio("324"))
print(valid_str_precio(".5"))
print(valid_str_precio("0003"))
print(valid_str_precio("54..4"))
print(valid_str_precio("precio"))
````

### Comentario
- Fíjate en un detalle curioso: `float("54..4")` ya lanza por sí sola un `ValueError` (dos puntos no forman un número decimal válido), así que el `except` ya devolvería `False` antes incluso de llegar a comprobar `cadena.count(".") <= 1`. Esa comprobación adicional no llega a ser necesaria en la práctica para este caso, pero no está de más como red de seguridad explícita, y ayuda a que la intención del código quede clara con solo leerlo.

## Resuelto 399: Eliminar las letras XYZ
* Crea una función que elimine las letras "x", "y" y "z" de la cadena dada y devuelva la versión modificada.
* Si la cadena dada no contiene "x", "y" o "z", devuelve `None`.

    2. remove_xyz("Dime lo que ves") ➞ None

````python
# Método 1: eliminando x, y, z
def remove_xyz(s):
    resultado = "".join(c for c in s if c.lower() not in "xyz")
    return resultado if resultado != s else None
````

````python
print(remove_xyz("Xilófono con zoom y rayos"))
print(remove_xyz("Dime lo que ves"))
print(remove_xyz(""))
````

### Comentario
- `resultado if resultado != s else None` compara el resultado ya filtrado con la cadena original: si no ha cambiado nada (es decir, ninguna letra "x", "y" o "z" ha sido eliminada), la cadena filtrada y la original son idénticas, y la función devuelve `None`. Comparar el resultado con el original antes y después de una transformación es una forma sencilla de detectar si esa transformación realmente hizo algo.

## Resuelto 400: Maratón de Cortes de Cadenas (2ª parte)
* Este desafío consta de cinco ejercicios en miniatura para practicar la habilidad de cortar cadenas.
* Echa un vistazo a los ejemplos a continuación para ver cómo se deben cortar las cadenas.
* ¡No se permite concatenar cadenas!
* Los resultados deben lograrse exclusivamente mediante el corte de cadenas.
* Ejemplos:
    - txt = "abcdefghijklmnopqrstuvwxyz"
    1. reto1(txt) ➞ "abcdefg"
        - Primeros 7 caracteres de la cadena.
    2. reto2(txt) ➞ "tuvwxyz"
        - Últimos 7 caracteres de la cadena.
    3. reto3(txt) ➞ "zyxwvutsrqponmlkjihgfedcba"
        - Todos los caracteres de la cadena en orden inverso.
    4. reto4(txt) ➞ "gfedcba"
        - Primeros 7 caracteres de la cadena (comenzando desde el séptimo carácter y retrocediendo).
    5. reto5(txt) ➞ "tvxz"
        - Tomar los últimos 8 caracteres y devolver solo los que están en posiciones impares.

````python
# Método 1: slicing: primeros 7
def reto1(s):
    return s[:7]

def reto2(s):
    return s[-7:]

def reto3(s):
    return s[::-1]

def reto4(s):
    return s[6::-1]

def reto5(s):
    return s[-8:][1::2]
````

````python
txt = "abcdefghijklmnopqrstuvwxyz"
print(reto1(txt))
print(reto2(txt))
print(reto3(txt))
print(reto4(txt))
print(reto5(txt))
````

### Comentario
- Este ejercicio es un buen repaso de las distintas variantes de slicing vistas a lo largo del curso: extremos con `[:n]` y `[-n:]`, inversión con `[::-1]`, e inversión parcial con paso negativo `[6::-1]` (equivalente a invertir los primeros 7 caracteres).
- El `reto5` encadena **dos slicings seguidos**: primero `s[-8:]` se queda con los últimos 8 caracteres, y después `[1::2]` se aplica sobre ese resultado (no sobre la cadena original) para quedarse con los caracteres en posiciones impares de ese fragmento. Encadenar slicings así es perfectamente válido: cada slice se aplica sobre lo que ha devuelto el anterior.
