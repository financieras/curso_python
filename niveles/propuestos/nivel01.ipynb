## Resuelto 1: Restar dos números
* Cree una función que tome dos números como argumentos y devuelva el resultado de restar el segundo al primero.
* Ejemplos:
    1. restar(9, 4) ➞ 5
    2. restar(3, 8) ➞ -5
    3. restar(10, 10) ➞ 0

````python
# Método 1: devolviendo el resultado directamente
def restar(a, b):
    return a - b
````

````python
print(restar(9, 4))
print(restar(3, 8))
print(restar(10, 10))
````

### Comentario
- Esta es la primera función del curso, así que repasemos su estructura:
  - `def` abre la definición de la función, seguida de su nombre y, entre paréntesis, sus **parámetros** (`a` y `b`).
  - El cuerpo de la función va indentado (con espacios) debajo de la línea `def`.
  - `return` termina la ejecución de la función y entrega el resultado a quien la llamó.
- Si fuera una suma `a + b` el orden de los parámetros no importaría.
- En este caso, se trata de una resta y el orden de los parámetros importa.
- No es lo mismo restar `a - b` que `b - a`.
- Esta propiedad (que el orden de los operandos no altere el resultado) se llama **conmutatividad**. La suma y el producto son conmutativos; la resta y la división, no.

## Resuelto 2: Convertir semanas a días
* Escriba una función que tome un número entero de `semanas` y las convierta a días.
* Ejemplos:
    1. semanas_a_dias(2) ➞ 14
    2. semanas_a_dias(4) ➞ 28
    3. semanas_a_dias(1) ➞ 7

````python
# Método 1: devolviendo el resultado directamente
def semanas_a_dias(semanas):
    return semanas * 7
````

````python
# Método 2: guardando el resultado antes de devolverlo
def semanas_a_dias(semanas):
    dias_totales = semanas * 7
    return dias_totales
````

````python
print(semanas_a_dias(2))
print(semanas_a_dias(4))
print(semanas_a_dias(1))
````

### Comentario
- Fíjate en los dos métodos: ambos hacen exactamente lo mismo.
- El **Método 1** devuelve el resultado directamente en una sola línea.
- El **Método 2** primero guarda el resultado en una variable con un nombre descriptivo (`dias_totales`) y luego lo devuelve.
- Ninguno es "más correcto" que el otro, pero usar una variable intermedia con buen nombre puede hacer el código más fácil de leer, sobre todo si el cálculo es complejo o si necesitas usar ese valor varias veces.

## Resuelto 3: Retornar el número anterior
* Cree una función que tome un número como argumento, lo decremente en 1 y devuelva el resultado.
* Ejemplos:
    1. restar_uno(5) ➞ 4
    2. restar_uno(0) ➞ -1
    3. restar_uno(-2) ➞ -3

````python
# Método 1: devolviendo el resultado directamente
def restar_uno(n):
    return n - 1
````

````python
# Método 2: guardando el resultado antes de devolverlo
def restar_uno(n):
    n = n - 1
    return n
````

````python
# Método 3: el operador -= es una forma abreviada de escribir lo mismo
def restar_uno(n):
    n -= 1    # equivale a: n = n - 1
    return n
````

````python
print(restar_uno(5))
print(restar_uno(0))
print(restar_uno(-2))
````

### Comentario
Operadores de asignación compuesta:
- n += 2 equivale a n = n + 2 (suma)
- n -= 2 equivale a n = n - 2 (resta)
- n *= 2 equivale a n = n * 2 (producto)
- n /= 2 equivale a n = n / 2 (división)
- n //= 2 equivale a n = n // 2 (división entera)
- n %= 2 equivale a n = n % 2 (módulo o resto)
- n \*\*= 2 equivale a n = n \*\* 2 (exponencial o potencias)

* Estos operadores no son exclusivos de los números: por ejemplo, con cadenas de texto o listas, `texto += "!"` añade el signo al final, y `lista += [4]` añade un elemento. Los iremos viendo en próximos ejercicios.

## Resuelto 4: Alquiler de moto
* Escriba una función que tome los `dias` y los `km` recorridos con una moto de alquiler y devuelva el coste total.
* Se cobran 20 € por día y 2 € por kilómetro.
* Ejemplos:
    1. alquiler_moto(3, 10) ➞ 80
    2. alquiler_moto(1, 0) ➞ 20
    3. alquiler_moto(7, 50) ➞ 240

````python
# Método 1: devolviendo el resultado directamente
def alquiler_moto(dias, km):
    return dias * 20 + km * 2
````

````python
# Método 2: guardando los resultados intermedios en variables con nombre
def alquiler_moto(dias, km):
    coste_dias = dias * 20
    coste_km = km * 2
    return coste_dias + coste_km
````

````python
print(alquiler_moto(3, 10))
print(alquiler_moto(1, 0))
print(alquiler_moto(7, 50))
````

### Comentario
- Fíjate en la expresión `dias * 20 + km * 2`: no lleva paréntesis y aun así el resultado es correcto.
- Esto es porque Python respeta el **orden de operaciones** habitual en matemáticas: primero las multiplicaciones y divisiones, después las sumas y restas.
- Si quisieras forzar otro orden, tendrías que usar paréntesis, igual que en matemáticas: `(dias + km) * 2` no sería lo mismo que `dias + km * 2`.

## Resuelto 5: El del medio
* Cree una función que reciba una lista de 3 elementos y devuelva el elemento del medio (el segundo).
* Ejemplos:
    1. el_del_medio([10, 20, 30]) ➞ 20
    2. el_del_medio(['lunes', 'martes', 'miércoles']) ➞ 'martes'
    3. el_del_medio([7, 15, 3]) ➞ 15

````python
# Método 1: indexando directamente la posición 1
def el_del_medio(lista):
    return lista[1]
````

````python
# Método 2: desempaquetando la lista en tres variables
def el_del_medio(lista):
    _, medio, _ = lista
    return medio
````

````python
print(el_del_medio([10, 20, 30]))
print(el_del_medio(['lunes', 'martes', 'miércoles']))
print(el_del_medio([7, 15, 3]))
````

### Comentario
El _indexing_ (en inglés) conocido como la __indexación__ o __acceso por índice__ de una lista se hace así:

`lista[n]` sirve para acceder directamente a **un único elemento** de la lista usando su número de posición (`n`), conocido como índice.

La regla clave es que Python empieza a contar **desde cero**:

* `lista[0]` es el **primer** elemento.
* `lista[1]` es el **segundo** elemento.
* `lista[2]` es el **tercer** elemento.

**Ejemplo:**

```python
colores = ["rojo", "verde", "azul"]

print(colores[0])  # "rojo"
print(colores[1])  # "verde"
print(colores[2])  # "azul"
```

* Índices negativos: si usas números negativos como `lista[-1]`, empiezas a contar desde el final, obteniendo el último elemento.

```python
print(colores[-1])  # "azul"
```

* Esto mismo funciona igual con las cadenas de texto (`str`), ya que también son secuencias: `"hola"[0]` es `"h"`.
* Más adelante veremos el _slicing_ (`lista[inicio:fin]`), que permite extraer **varios** elementos seguidos en lugar de uno solo.

## Resuelto 6: Velocidad media
* Cree una función que tome la `distancia` (en km) y el `tiempo` (en horas) y devuelva la velocidad media en km/h.
* Como el cálculo es una división, el resultado es un número decimal (`float`).
* Ejemplos:
    1. velocidad_media(100, 2) ➞ 50.0
    2. velocidad_media(30, 0.5) ➞ 60.0
    3. velocidad_media(10, 5) ➞ 2.0


````python
# Método 1: devolviendo el resultado directamente
def velocidad_media(distancia, tiempo):
    return distancia / tiempo
````

````python
print(velocidad_media(100, 2))
print(velocidad_media(30, 0.5))
print(velocidad_media(10, 5))
````

### Comentario
- En Python 3, el operador `/` (división) siempre devuelve un número decimal (`float`), aunque el resultado sea exacto.
- Por eso `velocidad_media(10, 5)` no devuelve `2`, sino `2.0`.
- Si en algún momento necesitas que el resultado sea un número entero (`int`), existe el operador de división entera `//`, que ya viste en el ejercicio 3 (por ejemplo, `10 // 5` sí devuelve `2`).

## Resuelto 8: Caramelos para los nietos

### Comentario

- Sirve para obtener el **residuo o resto** de una división entera.

- Por ejemplo, si divides 23 entre 5, el resultado es 4 y te sobran 3. El operador `%` te da ese 3.

```python
print(20 % 5)   # Módulo = 0 (división exacta)

print(22 % 5)   # Módulo = 2
print(23 % 5)   # Módulo = 3
print(24 % 5)   # Módulo = 4

print(26 % 5)   # Módulo = 1
```

* Un uso muy común: saber si un número es par o impar.
* Si `numero % 2 == 0`, el número es par.

## Resuelto 9: Media de notas de un profesor
* Un profesor anotó las calificaciones de dos exámenes como cadenas de texto y quiere calcular la media.
* Cree una función que reciba ambas notas como cadena y devuelva la media aritmética como número.
* Ejemplos:
    1. media_notas("7", "8") ➞ 7.5
    2. media_notas("10", "10") ➞ 10.0
    3. media_notas("5", "3") ➞ 4.0

````python
# Método 1: convirtiendo a entero en la misma expresión
def media_notas(nota1, nota2):
    return (int(nota1) + int(nota2)) / 2
````

````python
# Método 2: guardando las notas convertidas antes de calcular
def media_notas(nota1, nota2):
    n1 = int(nota1)
    n2 = int(nota2)
    return (n1 + n2) / 2
````

````python
print(media_notas("7", "8"))
print(media_notas("10", "10"))
print(media_notas("5", "3"))
````

### Comentario
- `int()` convierte una cadena de texto que representa un número entero en un verdadero número, para poder operar con él.
- Por ejemplo, `int("7")` convierte `"7"` en `7`.
- Si la cadena no representa un número válido (por ejemplo `int("siete")`), Python lanzará un error `ValueError`.
- Existe también `float()`, que convierte una cadena en un número decimal, por ejemplo `float("7.5")` ➞ `7.5`.

## Resuelto 12: ¿Has aprobado el examen?
* Un examen se aprueba con una nota mayor o igual a 5. Cree una función que reciba una nota y devuelva `True` si está aprobada, y `False` si no.
* Ejemplos:
    1. aprobado(7) ➞ True
    2. aprobado(4) ➞ False
    3. aprobado(5) ➞ True

````python
# Método 1: devolviendo la comparación directamente
def aprobado(nota):
    return nota >= 5
````

````python
# Método 2: con un condicional if/else
def aprobado(nota):
    if nota >= 5:
        return True
    else:
        return False
````

````python
print(aprobado(7))
print(aprobado(4))
print(aprobado(5))
````

### Comentario 1
- Las **condiciones lógicas** evalúan si una afirmación es verdadera o falsa.
- Al usar operadores de comparación (como `>=`), Python calcula el resultado y devuelve siempre uno de los dos **valores booleanos**:

* `True`: Significa verdadero. Indica que la condición se cumple.
* `False`: Significa falso. Indica que la condición NO se cumple.

- `True` y `False` pertenecen al tipo de dato `bool`. En Python, `bool` es en realidad un caso especial del tipo `int`: por dentro, `True` equivale a `1` y `False` equivale a `0` (lo veremos con más detalle en próximos ejercicios).
- Otros operadores de comparación habituales son `==` (igual), `!=` (distinto), `<`, `>`, `<=` y `>=`.

### Comentario 2
- Este es el primer **condicional** del curso, así que veamos su estructura básica:

```python
if condicion:
    # se ejecuta si la condición es True
    ...
else:
    # se ejecuta si la condición es False
    ...
```

- Python evalúa `condicion`. Si es `True`, ejecuta el bloque de debajo del `if`; si es `False`, ejecuta el bloque de debajo del `else`.
- Solo se ejecuta uno de los dos bloques, nunca ambos.
- Fíjate en la indentación (los espacios al principio de cada línea): en Python no se usan llaves `{}` como en otros lenguajes para delimitar los bloques; es la indentación la que indica qué instrucciones pertenecen a cada bloque.
- En este ejercicio concreto, como cada bloque termina con un `return`, el `else` podría omitirse (ya que si se cumple el `if`, la función termina ahí y nunca llega a la siguiente línea). Es una forma habitual y válida de escribir código más corto, que iremos viendo en próximos ejercicios; de momento, con `if`/`else` completo, se ve más claramente la estructura.

## Resuelto 14: Descuento especial
* Este es un caso de "código roto"
* Una tienda aplica un descuento del 10 % solo a clientes VIP. Pero el programador novato lo hizo mal.
* **Código roto:**
  ```python
  def precio_final(precio, es_vip):
      return precio * 0.90
      if es_vip:
          return precio * 0.90
      return precio
  ```
* Ejemplos:
    1. precio_final(100, False) ➞ 100
    2. precio_final(100, True) ➞ 90.0
    3. precio_final(50, True) ➞ 45.0


````python
# Método 1: corrigiendo con un condicional if/else
def precio_final(precio, es_vip):
    if es_vip:
        return precio * 0.90
    return precio
````

````python
# Método 2: con un operador ternario
def precio_final(precio, es_vip):
    return precio * 0.90 if es_vip else precio
````

````python
print(precio_final(100, False))
print(precio_final(100, True))
print(precio_final(50, True))
````

### Comentario 1
- El título dice "condicional if/else", pero fíjate en que el código no tiene ningún `else`:

```python
def precio_final(precio, es_vip):
    if es_vip:
        return precio * 0.90
    return precio
```

- Esto es válido porque, en cuanto se ejecuta un `return` dentro del `if`, la función **termina inmediatamente** y no llega a ejecutar ninguna línea posterior.
- Por tanto, `return precio` solo se ejecuta cuando `es_vip` es `False` (si fuera `True`, la función ya habría terminado en la línea anterior). Esa última línea cumple exactamente la misma función que un `else`, sin necesidad de escribirlo.
- Es una forma muy habitual de simplificar un `if`/`else` en Python cuando ambos bloques terminan en `return`. Equivale exactamente a:

```python
def precio_final(precio, es_vip):
    if es_vip:
        return precio * 0.90
    else:
        return precio
```

### Comentario 2
- El **Método 2** introduce el **operador ternario** (u operador condicional en línea), que sigue esta forma general:

```python
A if condicion else B
```

- Se lee así:
    - Si `condicion` es `True`, se devuelve `A`; en caso contrario, se devuelve `B`
- Es una forma compacta de escribir un `if`/`else` que solo decide **qué valor devolver**, en una única línea, en lugar de repartir la lógica en varias líneas como en un `if`/`else` tradicional.

## Resuelto 16: Volumen de un cubo
* Cree una función que reciba el `lado` de un cubo y devuelva su volumen.
* La fórmula es lado ** 3.
* Ejemplos:
    1. volumen_cubo(3) ➞ 27
    2. volumen_cubo(5) ➞ 125
    3. volumen_cubo(10) ➞ 1000

````python
# Método 1: con el operador **
def volumen_cubo(lado):
    return lado ** 3
````

````python
# Método 2: con la función pow()
def volumen_cubo(lado):
    return pow(lado, 3)
````

````python
print(volumen_cubo(3))
print(volumen_cubo(5))
print(volumen_cubo(10))
````

### Comentario
- El operador `**` calcula potencias: `base ** exponente`.
- Por ejemplo, `lado ** 3` es "lado al cubo" (`lado * lado * lado`).

## Resuelto 23: Juntar dos equipos
* Cree una función que reciba dos listas de nombres de equipos y devuelva una sola lista con todos los miembros.
* Ejemplos:
    1. unir_equipos(['Ana', 'Luis'], ['Pedro', 'María']) ➞ ['Ana', 'Luis', 'Pedro', 'María']
    2. unir_equipos(['Sofía'], []) ➞ ['Sofía']
    3. unir_equipos([], []) ➞ []

````python
# Método 1: concatenando listas con +
def unir_equipos(eq1, eq2):
    return eq1 + eq2
````

````python
# Método 2: usando el método extend()
def unir_equipos(eq1, eq2):
    resultado = eq1.copy()
    resultado.extend(eq2)
    return resultado
````

````python
# Método 3: recorriendo eq2 y añadiendo cada elemento con append()
def unir_equipos(eq1, eq2):
    resultado = eq1.copy()
    for miembro in eq2:
        resultado.append(miembro)
    return resultado
````

````python
print(unir_equipos(['Ana', 'Luis'], ['Pedro', 'María']))
print(unir_equipos(['Sofía'], []))
print(unir_equipos([], []))
````

### Comentario
- `eq1 + eq2` con listas no sirve para sumar números: **concatena** ambas listas en una nueva, uniendo sus elementos uno detrás de otro.
- `lista.copy()` crea una copia independiente de la lista, para no modificar accidentalmente la original al añadirle elementos.
- `resultado.extend(eq2)` añade, al final de `resultado`, **cada elemento** de `eq2` por separado. Es equivalente a recorrer `eq2` con un bucle `for` y usar `resultado.append(miembro)` en cada vuelta, como en el Método 3.
- Ojo con la diferencia: si en vez de `extend(eq2)` hubiéramos escrito `resultado.append(eq2)`, `eq2` se habría añadido **entera y de una sola vez**, como un único elemento (una lista dentro de la lista), en lugar de añadir sus miembros uno a uno. Por ejemplo, `resultado.append(['Pedro', 'María'])` daría `[..., ['Pedro', 'María']]`, no `[..., 'Pedro', 'María']`.

- `dias[:n]` es un ejemplo de **slicing** (rebanado): a diferencia de `dias[n]`, que devuelve un único elemento, `dias[:n]` devuelve una **nueva lista** con los primeros `n` elementos.
- La sintaxis general es `lista[inicio:fin]`, donde se incluye el elemento en la posición `inicio` pero **no** el de la posición `fin`.
- Si se omite `inicio` (como aquí), se empieza desde el principio de la lista; si se omite `fin`, se llega hasta el final.

- `type(n) is int` comprueba el tipo exacto del valor: como en Python `bool` es un subtipo de `int` (Resuelto 17), pero `type()` no tiene en cuenta esa herencia, `type(True) is int` da `False` directamente (el tipo exacto de `True` es `bool`, no `int`).
- Por eso, en el Método 1, el trozo `and type(n) is not bool` es en realidad **redundante**: si `type(n) is int` ya es `False`, el resultado final ya es `False` sin necesidad de mirar nada más; y si `type(n) is int` es `True`, `n` no puede ser al mismo tiempo de tipo `bool` (un valor no puede tener dos tipos exactos a la vez), así que esa segunda comprobación nunca cambia el resultado. El Método 1 funciona igual de bien escrito solo como `return type(n) is int`.
- La cosa cambia con `isinstance()` (Método 2): como `isinstance()` sí tiene en cuenta la herencia, `isinstance(True, int)` da `True` (porque `bool` hereda de `int`). Por eso aquí `and not isinstance(n, bool)` no es redundante, sino necesario para excluir los booleanos.
- En general `isinstance()` es preferible a `type() is` porque reconoce subclases (útil, por ejemplo, si se crean tipos numéricos personalizados); pero esa misma característica es la que obliga, en este caso concreto, a excluir `bool` explícitamente solo con `isinstance()`, no con `type()`.

## Resuelto 43: Código de acceso capicúa

* Un sistema de seguridad proporciona acceso preferente si el código es "capicúa": se lee igual de izquierda a derecha que de derecha a izquierda.
* Cree una función que reciba un código (como cadena de texto) y devuelva `True` si es capicúa; en caso contrario, `False`.
* Ejemplos:
    1. codigo_capicua("2002") ➞ True
    2. codigo_capicua("radar") ➞ True
    3. codigo_capicua("1234") ➞ False
    4. codigo_capicua("a") ➞ True

````python
# Método 1: comparando con la cadena invertida usando slicing
def codigo_capicua(codigo):
    return codigo == codigo[::-1]
````

````python
# Método 2: usando la función reversed()
def codigo_capicua(codigo):
    return codigo == "".join(reversed(codigo))
````

````python
print(codigo_capicua("2002"))
print(codigo_capicua("radar"))
print(codigo_capicua("1234"))
print(codigo_capicua("a"))
````

### Comentario
- `codigo[::-1]` es un slicing con **paso** (`step`) igual a `-1`: recorre la cadena de atrás hacia adelante, es decir, la invierte.
- La sintaxis completa de un slice es `secuencia[inicio:fin:paso]`; al dejar `inicio` y `fin` vacíos y poner `paso = -1`, se obtiene la secuencia entera invertida.
- `reversed(codigo)` hace algo parecido, pero devuelve un objeto iterable (no una cadena), por eso hay que unirlo con `"".join(...)` para reconstruir el texto.

- Al llamar a `random.` necesitas escribir primero el nombre del módulo (`random`) seguido de un punto y el nombre de la función (`randint`), porque `randint` "vive dentro" del módulo `random`.
- `"5" in id_str` comprueba si el carácter `"5"` aparece en algún punto de la cadena `id_str`. `id_str.endswith("0")` comprueba si la cadena termina en `"0"`.

````python
# Método 1: eliminando el mínimo y el máximo con remove(), luego calculando la media
def media_central(lista):
    if len(lista) < 3:
        return None
    copia = lista.copy()
    copia.remove(min(copia))
    copia.remove(max(copia))
    return sum(copia) / len(copia)
````

````python
# Método 2: ordenando la lista y descartando los extremos
def media_central(lista):
    if len(lista) < 3:
        return None
    ordenada = sorted(lista)
    centro = ordenada[1:-1]
    return sum(centro) / len(centro)
````

````python
print(media_central([10, 20, 30, 40, 50, 50]))
print(media_central([5, 5, 5, 5, 3]))
print(media_central([1, 100]))
print(media_central([]))
````

### Comentario
- `lista.remove(valor)` busca la **primera** aparición de ese valor en la lista y la elimina (a diferencia de `lista.pop(indice)`, que elimina por posición). Si el valor aparece varias veces, solo se elimina una ocurrencia, que es justo lo que pide este ejercicio.
- `ordenada[1:-1]` es un slicing que descarta el primer y el último elemento de la lista ya ordenada, dejando solo los del "centro".

## Resuelto 96: Central de alarmas
* Una central de alarmas recibe las señales de un número variable de sensores. Recuerde que un número indeterminado de parámetros se recoge con la sintaxis **\*args**.
* Si al menos una señal es verdadera, hay que dar la alerta.
* Los argumentos pueden ser cualquier tipo de dato en Python: se evalúan como booleanos.
* Escriba una función llamada `hay_alerta` que devuelva `True` si algún argumento es verdadero y `False` si ninguno lo es.
* Ejemplos:
    1. hay_alerta(False, False, True) ➞ True
    2. hay_alerta(0, "", []) ➞ False
    3. hay_alerta("humo") ➞ True
    4. hay_alerta(None, 0, False) ➞ False

````python
# Método 1: usando any()
def hay_alerta(*args):
    return any(args)
````

````python
# Método 2: recorriendo los argumentos manualmente
def hay_alerta(*args):
    for arg in args:
        if arg:
            return True
    return False
````

````python
print(hay_alerta(False, False, True))
print(hay_alerta(0, "", []))
print(hay_alerta("humo"))
print(hay_alerta(None, 0, False))
````

### Comentario 1

En Python, todos los objetos tienen un valor booleano intrínseco cuando se evalúan en un contexto lógico (como un `if` o al usar funciones lógicas). Esto se conoce como valores **Truthy** (verdaderos) y **Falsy** (falsos):

* **Valores Falsy (evalúan como `False`):** `False`, `None`, el cero (`0`, `0.0`), y cualquier colección o cadena vacía (`""`, `[]`, `{}`, `()`).
* **Valores Truthy (evalúan como `True`):** Cualquier cosa que no sea Falsy. Esto incluye números distintos de cero y **cadenas de texto que no estén vacías**.

Como `"humo"` es una cadena de texto con contenido (no está vacía), Python la evalúa como **`True`**.

### Comentario 2
- `*args` en la definición de una función permite recibir un **número indeterminado de argumentos posicionales**: dentro de la función, `args` es una tupla con todos los valores recibidos, sin importar cuántos sean.
- Por eso `hay_alerta(False, False, True)` y `hay_alerta("humo")` funcionan con la misma función, aunque se llamen con distinto número de argumentos.
- `any(args)` devuelve `True` si al menos un elemento de esa tupla es "truthy", aprovechando justo los conceptos de Truthy/Falsy explicados arriba.

## Resuelto 97: Racha de aprobados
* Un estudiante apunta sus notas en orden cronológico y quiere presumir de su racha inicial de aprobados.
* Escriba una función llamada `racha_aprobados` que reciba la lista de notas y devuelva otra lista con las notas desde el principio hasta el primer suspenso (nota menor que 5).
* Al llegar a un suspenso, no se incluye ni él ni nada de lo posterior.
* Ejemplos:
    1. racha_aprobados([7, 6, 9, 4, 8]) ➞ [7, 6, 9]
    2. racha_aprobados([3, 7, 10]) ➞ []
    3. racha_aprobados([5, 5, 10]) ➞ [5, 5, 10]


````python
# Método 1: recorriendo la lista con break
def racha_aprobados(notas):
    resultado = []
    for nota in notas:
        if nota < 5:
            break
        resultado.append(nota)
    return resultado
````

````python
# Método 2: con un bucle while
def racha_aprobados(notas):
    resultado = []
    i = 0
    while i < len(notas) and notas[i] >= 5:
        resultado.append(notas[i])
        i += 1
    return resultado
````

````python
print(racha_aprobados([7, 6, 9, 4, 8]))
print(racha_aprobados([3, 7, 10]))
print(racha_aprobados([5, 5, 10]))
````

### Comentario 1
- `break` es una instrucción que **interrumpe inmediatamente** un bucle `for` (o `while`), sin esperar a que termine de recorrer toda la secuencia.
- En este caso, en cuanto se encuentra la primera nota suspensa (`nota < 5`), el bucle se detiene con `break` y ya no se examinan las notas posteriores, por lo que tampoco se añaden al resultado.
- `break` es útil precisamente cuando la condición de parada **no forma parte natural** de la cabecera del bucle: aquí recorremos "nota por nota" y decidimos, dentro del bucle, si merece la pena seguir. Compáralo con el Método 2, donde el bucle `while` no necesita `break` porque la condición de parada se puede expresar directamente en su cabecera.

### Comentario 2
- El **Método 2** introduce el bucle **`while`**, que repite un bloque de instrucciones **mientras se cumpla una condición**, sin saber de antemano cuántas vueltas hará falta.

```python
while condicion:
    # este bloque se repite una y otra vez, mientras "condicion" sea True
    ...
```

- A diferencia del `for` (pensado para recorrer una secuencia conocida, elemento a elemento), el `while` es la herramienta natural cuando el número de repeticiones depende de algo que puede cambiar en cada vuelta.
- En este ejercicio, la condición `i < len(notas) and notas[i] >= 5` expresa **todo** el criterio de parada de golpe: "mientras queden notas por revisar Y la nota actual sea un aprobado". En cuanto una de las dos deja de cumplirse, el bucle termina solo, sin necesidad de un `break` explícito.
- Nota importante: aunque aquí no se use, `while` también admite `break` (igual que el `for` del Método 1), para los casos en que la condición de parada sea más fácil de detectar *dentro* del bucle que de escribir en su cabecera.
- No olvides actualizar `i` dentro del bucle (`i += 1`): si no lo haces, la condición nunca cambiará y el bucle se repetirá **para siempre** (lo que se conoce como un bucle infinito).

## Resuelto 98: Cola de impresión
* Una impresora atiende los documentos por orden: el primero de la lista es el siguiente en imprimirse.
* Cree una función que devuelva el primer valor de una lista o de una cadena.
* Las listas o las cadenas pueden ser de diferentes tamaños.
* Devuelva `None` si la lista o la cadena está vacía.
* Ejemplos:
    1. siguiente(["informe.pdf", "foto.png", "acta.txt"]) ➞ "informe.pdf"
    2. siguiente("Python") ➞ "P"
    3. siguiente([]) ➞ None
    4. siguiente("") ➞ None


````python
# Método 1: comprobando si está vacío con un condicional
def siguiente(objeto):
    if not objeto:
        return None
    return objeto[0]
````

````python
print(siguiente(["informe.pdf", "foto.png", "acta.txt"]))
print(siguiente("Python"))
print(siguiente([]))
print(siguiente(""))
````

## Resuelto 99: Vueltas rápidas
* En una carrera de karts se registra el tiempo (en segundos) de cada vuelta que da un piloto.
* La "vuelta rápida" es la de menor tiempo de toda la carrera.
* Cree una función que reciba la lista de tiempos y devuelva cuántas vueltas igualaron ese mejor tiempo.
* Ejemplos:
    1. vueltas_rapidas([62, 60, 60, 65]) ➞ 2
    2. vueltas_rapidas([70]) ➞ 1
    3. vueltas_rapidas([59, 59, 59]) ➞ 3
    4. vueltas_rapidas([61, 58, 63, 58]) ➞ 2


````python
# Método 1: usando min() y count()
def vueltas_rapidas(tiempos):
    vuelta_rapida = min(tiempos)
    return tiempos.count(vuelta_rapida)
````

````python
print(vueltas_rapidas([62, 60, 60, 65]))
print(vueltas_rapidas([70]))
print(vueltas_rapidas([59, 59, 59]))
print(vueltas_rapidas([61, 58, 63, 58]))
````

### Comentario
- `lista.count(valor)` recorre una lista y devuelve cuántas veces aparece exactamente ese valor. Es el equivalente, para listas, al método `.count()` que ya usamos en cadenas de texto (Resuelto 100).

## Resuelto 100: ¡Gol, gol, gol!
* Un periodista deportivo quiere saber cuántas veces gritó "gol" el narrador en la transcripción de un partido.
* Cree una función que devuelva la cantidad de veces que aparece "gol" en una cadena.
* Ejemplos:
    1. cuenta_goles("golgolgol") ➞ 3
    2. cuenta_goles("¡golazo del delantero!") ➞ 1
    3. cuenta_goles("hoy empate sin emoción") ➞ 0
    4. cuenta_goles("gol y gol y otro gol") ➞ 3


````python
# Método 1: usando el método count()
def cuenta_goles(cadena):
    return cadena.count("gol")
````

````python
print(cuenta_goles("golgolgol"))
print(cuenta_goles("¡golazo del delantero!"))
print(cuenta_goles("hoy empate sin emoción"))
print(cuenta_goles("gol y gol y otro gol"))
````

### Comentario
- `cadena.count(subcadena)` cuenta cuántas veces aparece `subcadena` dentro de `cadena`, sin solaparse: por ejemplo, `"aaa".count("aa")` da `1`, no `2`, porque una vez usados los dos primeros caracteres para la primera coincidencia, solo queda uno para buscar la siguiente.
