## Resuelto 101: Funciones Lambda
* Una tienda online cobra un gasto de envío del 5% sobre el subtotal de la compra.
* Escriba una función llamada `calcular_coste_total` que tome dos argumentos: `precio_unitario` y `cantidad`.
* La función debe calcular el subtotal, el gasto de envío y el coste total, y devolver los tres valores como una tupla `(subtotal, gastos_envio, total)`.
* Use como función principal `calcular_coste_total(precio_unitario, cantidad)` y dentro de ella llame a una función secundaria `calcular_gastos_envio(subtotal, porcentaje)`.
* En el Método 1 use dos funciones, donde la principal llama a la secundaria.
* En el Método 2 use una función lambda dentro de la función principal.
* Ejemplos:
    1. calcular_coste_total(10, 4) ➞ (40, 2.0, 42.0)
    2. calcular_coste_total(25, 2) ➞ (50, 2.5, 52.5)
    3. calcular_coste_total(100, 1) ➞ (100, 5.0, 105.0)


````python
# Método 1: con una función auxiliar
def calcular_coste_total(precio_unitario, cantidad):
    def calcular_gastos_envio(subtotal, porcentaje):
        return subtotal * porcentaje / 100
    subtotal = precio_unitario * cantidad
    gastos_envio = calcular_gastos_envio(subtotal, 5)
    total = subtotal + gastos_envio
    return (subtotal, gastos_envio, total)
````

````python
# Método 2: con una función lambda dentro de la principal
def calcular_coste_total(precio_unitario, cantidad):
    calcular_gastos_envio = lambda subtotal, porcentaje: subtotal * porcentaje / 100
    subtotal = precio_unitario * cantidad
    gastos_envio = calcular_gastos_envio(subtotal, 5)
    total = subtotal + gastos_envio
    return (subtotal, gastos_envio, total)
````

````python
print(calcular_coste_total(10, 4))
print(calcular_coste_total(25, 2))
print(calcular_coste_total(100, 1))
````

### Comentario
- Este ejercicio introduce una **función anidada**: `calcular_gastos_envio` se define *dentro* de `calcular_coste_total`, por lo que solo existe y puede usarse dentro de esa función.
- El Método 2 hace lo mismo pero con una función lambda en vez de una función con `def`. Ambos métodos son equivalentes: la diferencia es solo de estilo, ya que la lógica interna (`subtotal * porcentaje / 100`) es igual de simple en los dos casos.
- Fíjate en que la función `return (subtotal, gastos_envio, total)` devuelve **tres valores a la vez** empaquetados en una tupla, como ya vimos en el Resuelto 81 del Nivel 1.

## Resuelto 104: ¿La despensa tiene ese ingrediente?
* Escriba una función que devuelva `True` si un diccionario (la despensa) contiene la clave especificada (el ingrediente) y `False` en caso contrario.
* Ejemplos:
    1. tiene_ingrediente({'harina': 2, 'huevos': 6, 'leche': 1}, "azúcar") ➞ False
    2. tiene_ingrediente({'sal': True, 'pimienta': True, 'aceite': True}, "vinagre") ➞ False
    3. tiene_ingrediente({'ene': 1, 'feb': 2, 'mar': 3}, "feb") ➞ True


````python
# Método 1: usando el operador in en el diccionario
def tiene_ingrediente(despensa, ingrediente):
    return ingrediente in despensa
````

````python
# Método 2: usando el método get()
def tiene_ingrediente(despensa, ingrediente):
    return despensa.get(ingrediente) is not None
````

````python
print(tiene_ingrediente({'harina': 2, 'huevos': 6, 'leche': 1}, "azúcar"))
print(tiene_ingrediente({'sal': True, 'pimienta': True, 'aceite': True}, "vinagre"))
print(tiene_ingrediente({'ene': 1, 'feb': 2, 'mar': 3}, "feb"))
````

### Comentario
- El Método 1 (`ingrediente in despensa`) comprueba directamente si esa clave existe en el diccionario, sin importar qué valor tenga asociado.
- El Método 2 (`despensa.get(ingrediente) is not None`) funciona bien en los ejemplos de este ejercicio, pero tiene una **limitación**: si algún ingrediente estuviera guardado con el valor `None` (por ejemplo, `{'sal': None}`), `despensa.get('sal')` devolvería `None`, y la función diría erróneamente que la despensa NO tiene ese ingrediente, aunque la clave sí existiera.
- Por eso, para comprobar si una clave existe (sin importar su valor), el operador `in` (Método 1) es, en general, más fiable que comparar el resultado de `.get()` con `None`.

## Resuelto 105: Tormenta de truenos
* Durante una tormenta se oyen truenos que se solapan entre sí, como en 'TruTruTrun!' (el primer trueno se corta y en total hay 3 truenos).
* Dada una serie de cómo sonaron los truenos superpuestos, indica cuántos truenos se oyeron en total.
* Ejemplos:
    1. contar_truenos("TruTruTrun!") ➞ 3
    2. contar_truenos("TrTrTruTruTruTrun!") ➞ 6
    3. contar_truenos("TTTruTrTrun!Trun!TrTrTrun!") ➞ 9


````python
# Método 1: usando count() sobre la cadena
def contar_truenos(cadena):
    return cadena.count("T")
````

````python
# Método 2: recorriendo la cadena por bloques de 3
def contar_truenos(cadena):
    contador = 0
    for c in cadena:
        if c == "T":
            contador += 1
    return contador
````

````python
print(contar_truenos("TruTruTrun!"))
print(contar_truenos("TrTrTruTruTruTrun!"))
print(contar_truenos("TTTruTrTrun!Trun!TrTrTrun!"))
````

### Comentario
- Fíjate en que ambos métodos simplemente cuentan cuántas veces aparece la letra `"T"` en la cadena, sin fijarse en los bloques de tres letras ("Tru").
- Esto funciona porque cada trueno, se solape o no con otro, siempre aporta **exactamente una** letra `"T"` a la cadena final (es la letra con la que empieza "Tru"); las letras compartidas por el solapamiento son las que vienen después (`"ru"`), nunca la `"T"` inicial. Por eso, contar las "T" equivale a contar los truenos, sin necesidad de analizar bloques de tres en tres.

## Resuelto 106: Engranajes
* En un mecanismo simple hay tres piezas: el engranaje motriz "m", el engranaje intermedio "i" y el engranaje conducido "c".
* Según cuál de las tres piezas se sitúe en el centro de la lista, el mecanismo se clasifica en tipo A, tipo B o tipo C:
1. Si en el centro está el engranaje conducido "c", es un mecanismo de tipo A.
2. Si en el centro está el engranaje motriz "m", es un mecanismo de tipo B.
3. Si en el centro está el engranaje intermedio "i", es un mecanismo de tipo C.
* Dada una lista con las tres piezas en algún orden, escriba una función que determine el tipo de mecanismo.
* Ejemplos:
    1. tipo_mecanismo(['i', 'c', 'm']) ➞ "tipo A"
    2. tipo_mecanismo(['i', 'm', 'c']) ➞ "tipo B"
    3. tipo_mecanismo(['c', 'i', 'm']) ➞ "tipo C"


````python
# Método 1: inspeccionando el elemento central con un diccionario
def tipo_mecanismo(lista):
    centro = lista[1]
    tipos = {"c": "tipo A", "m": "tipo B", "i": "tipo C"}
    return tipos[centro]
````

````python
# Método 2: con condicionales if/elif
def tipo_mecanismo(lista):
    centro = lista[1]
    if centro == "c":
        return "tipo A"
    elif centro == "m":
        return "tipo B"
    return "tipo C"
````

````python
print(tipo_mecanismo(['i', 'c', 'm']))
print(tipo_mecanismo(['i', 'm', 'c']))
print(tipo_mecanismo(['c', 'i', 'm']))
````

## Resuelto 107: Redundancia booleana
* Este es un caso de "código roto".
* Disponemos de un código con algún pequeño error y, además, poco elegante.
* Su objetivo es presentar una función lo más minimalista posible.
* Escriba una función que devuelva la palabra "mayor" si la edad dada es mayor o igual a 18, y la palabra "menor" en caso contrario.
* Recuerde: comparar una condición con `== True` o `== False` es redundante, ya que la condición ya es un booleano por sí misma.
* Código erróneo:

```python
def clasificar_edad(años):
	resp = bool(años >= 18)
	if res == False:
		return "menor"
	if res == True:
		return "mayor"
```

* Ejemplos:
    1. clasificar_edad(20) ➞ "mayor"
    2. clasificar_edad(15) ➞ "menor"
    3. clasificar_edad(18) ➞ "mayor"


````python
# Método 1: con un ternario, sin redundancia booleana
def clasificar_edad(edad):
    return "mayor" if edad >= 18 else "menor"
````

````python
# Método 2: con un condicional if/else
def clasificar_edad(edad):
    if edad >= 18:
        return "mayor"
    return "menor"
````

````python
print(clasificar_edad(20))
print(clasificar_edad(15))
print(clasificar_edad(18))
````

### Comentario
- El código roto tenía dos problemas típicos:
  - Una **errata en el nombre de una variable**: se define `resp`, pero luego se usa `res` (sin la "p"). Python no las reconoce como la misma variable, así que este código daría un error `NameError: name 'res' is not defined`.
  - Una **redundancia booleana**: `bool(años >= 18)` ya es innecesario, porque `años >= 18` ya es, por sí mismo, un valor `bool` (`True` o `False`); envolverlo en `bool(...)` no cambia nada. Y comparar después con `== True` o `== False` es redundante por la misma razón: si `x` ya es un booleano, basta con escribir `if x:` o `if not x:`, sin compararlo explícitamente.
- El Método 1 evita ambos problemas usando directamente un operador ternario sobre la condición original.

```python
def puede_despegar(bateria_suficiente):
```

````python
# Método 2: con un operador ternario
def puede_despegar(bateria_suficiente):
    return "Despegue autorizado" if bateria_suficiente else "Despegue cancelado"
````

- Ambos son errores muy habituales al empezar: uno de identificadores (no pueden llevar espacios) y otro de la estructura del propio lenguaje (`else` no admite condiciones).

## Resuelto 110: Capacidad del ascensor
* Cree una función que devuelva `True` si el peso combinado de un ascensor `peso_ascensor` y el peso de las personas `p` que suben es menor que el peso máximo `peso_maximo` que puede soportar. De lo contrario, devuelve `False`.
* El peso del ascensor y el peso de las personas se dan en libras.
* El peso máximo se indica en kilogramos.
* Nota: 1 libra = 0,453592 kilogramo
* El orden de los argumentos en la función son: `capacidad_permitida(peso_ascensor, p, peso_maximo)`
* Ejemplos:
    1. capacidad_permitida(200, [150, 160, 140], 400) ➞ True
    2. capacidad_permitida(300, [220, 210, 205], 350) ➞ False
    3. capacidad_permitida(150, [130, 120], 200) ➞ True


````python
# Método 1: convirtiendo libras a kilos y comparando
def capacidad_permitida(peso_ascensor, p, peso_maximo):
    peso_total_lb = peso_ascensor + sum(p)
    peso_total_kg = peso_total_lb * 0.453592
    return peso_total_kg < peso_maximo
````

````python
# Método 2: con sum() en una sola línea
def capacidad_permitida(peso_ascensor, p, peso_maximo):
    return 0.453592 * (peso_ascensor + sum(p)) < peso_maximo
````

````python
print(capacidad_permitida(200, [150, 160, 140], 400))
print(capacidad_permitida(300, [220, 210, 205], 350))
print(capacidad_permitida(150, [130, 120], 200))
````

## Resuelto 113: Ligeramente por delante
* Se le darán dos listas muy parecidas o exactamente iguales, que representan las vueltas de dos corredores en una carrera.
* Si son exactamente iguales la función debe devolver `False`.
* Si difieren en un único elemento y éste es mayor en la primera lista se retornará `True`, y si es mayor el de la segunda lista se retornará `False`.
* Nunca se recibirán dos listas en las que puedan diferir más de un elemento.
* Ambas listas tendrán la misma longitud, aunque ésta puede ser variable.
* Ejemplos:
    1. va_por_delante([1, 2, 4], [1, 2, 3]) ➞ True
    2. va_por_delante([10, 20, 25, 40], [10, 20, 30, 40]) ➞ False
    3. va_por_delante([1, 2, 3, 4], [1, 2, 4, 4]) ➞ False
    4. va_por_delante([5, 5, 5], [5, 5, 5]) ➞ False


````python
# Método 1: comparando elemento a elemento
def va_por_delante(lst1, lst2):
    if lst1 == lst2:
        return False
    for i in range(len(lst1)):
        if lst1[i] != lst2[i]:
            return lst1[i] > lst2[i]
    return False
````

````python
print(va_por_delante([1, 2, 4], [1, 2, 3]))
print(va_por_delante([10, 20, 25, 40], [10, 20, 30, 40]))
print(va_por_delante([1, 2, 3, 4], [1, 2, 4, 4]))
print(va_por_delante([5, 5, 5], [5, 5, 5]))
````

### Comentario
- `range(len(lst1))` genera los índices `0, 1, 2, ...` hasta el último índice válido de `lst1`, y se usa para recorrer **dos listas a la vez, posición a posición** (`lst1[i]` y `lst2[i]`).
- En cuanto se detecta la primera posición donde ambas listas difieren, la función devuelve el resultado de esa comparación inmediatamente con `return`, sin seguir revisando el resto.
- Más adelante (Resuelto 138) veremos `zip()`, una herramienta que permite recorrer dos listas en paralelo de una forma todavía más directa, sin tener que usar índices.

* Ejemplos:
    1. saldos_opuestos([100, 50, 30, 20, 10]) ➞ [-100, -50, -30, -20, -10]
    2. saldos_opuestos([50, -20, 30, -40, 0]) ➞ [-50, 20, -30, 40, 0]

## Resuelto 127: Métodos Is...
* Este es un caso de "código roto".
* Queremos eliminar las letras de una matrícula y quedarnos solo con los números.
* Tenemos un código que no funciona correctamente para superar este desafío.


    3. solo_numeros("A1B2C3") ➞ "123"


````python
# Método 1: filtrando solo dígitos con isdigit()
def solo_numeros(cadena):
    return "".join(c for c in cadena if c.isdigit())
````

- `isdigit()` forma parte de una familia de métodos de cadena que empiezan por "is" y sirven para comprobar el tipo de contenido de un texto, devolviendo siempre `True` o `False`. Algunos de los más usados son:
  - `.isdigit()`: ¿son todo dígitos?
  - `.isalpha()`: ¿son todo letras?
  - `.isalnum()`: ¿son todo letras o dígitos (alfanumérico)?
  - `.isupper()` / `.islower()`: ¿está todo en mayúsculas / minúsculas?
  - `.isspace()`: ¿son todo espacios en blanco?

## Resuelto 133: Encontrar el error en un examen
* Cree una función para encontrar el valor `None` en una lista de notas de un examen (una nota `None` significa que ese alumno no se presentó).
* El valor de retorno debe ser el índice donde se encuentre `None`.
* Si no se encuentra `None` en la lista, la función devolverá -1.
* Ejemplos:
    1. encontrar_ausente([8, 6, None]) ➞ 2
    2. encontrar_ausente([None, 5, 7, 9]) ➞ 0
    3. encontrar_ausente([4, 5, 6, 7]) ➞ -1


````python
# Método 1: usando el método index() con try/except
def encontrar_ausente(lista):
    try:
        return lista.index(None)
    except ValueError:
        return -1
````

````python
# Método 2: con un bucle
def encontrar_ausente(lista):
    for i, nota in enumerate(lista):
        if nota is None:
            return i
    return -1
````

````python
# Método 3: con un ternario y 'if None in'
def encontrar_ausente(lista):
    return lista.index(None) if None in lista else -1
````

````python
print(encontrar_ausente([8, 6, None]))
print(encontrar_ausente([None, 5, 7, 9]))
print(encontrar_ausente([4, 5, 6, 7]))
````

### Comentario
- Este ejercicio introduce dos herramientas nuevas e importantes:

**`try` / `except`** (Método 1): permite "intentar" ejecutar un código que podría fallar, y capturar el error si ocurre, en lugar de que el programa se detenga:

```python
try:
    # código que podría lanzar un error
    ...
except ValueError:
    # se ejecuta solo si ese código lanza un error de tipo ValueError
    ...
```

- Aquí, `lista.index(None)` busca la posición de `None` en la lista, pero **lanza un error `ValueError`** si `None` no está presente. En vez de dejar que el programa se detenga con ese error, lo "atrapamos" con `except ValueError` y devolvemos `-1`.

**`enumerate()`** (Método 2): al recorrer una lista con `for nota in lista:`, solo tienes el valor de cada elemento, no su posición. `enumerate(lista)` te da **ambos a la vez**: el índice y el valor, en cada vuelta del bucle:

```python
for i, nota in enumerate(lista):
    # i es la posición (0, 1, 2, ...), nota es el valor en esa posición
    ...
```

- Fíjate también en `nota is None` (en vez de `nota == None`): para comparar con `None`, la forma recomendada en Python es usar `is` (que comprueba si es exactamente el mismo objeto `None`), en lugar de `==`. Con `None` ambas formas casi siempre dan el mismo resultado, pero `is None` es la práctica estándar y evita sorpresas en casos especiales.

## Resuelto 146: Suma de máximos en una matriz
* Cree una función que tome una matriz y devuelva la suma del valor máximo en cada fila.
* Una matriz es una lista de listas, esto es, una lista donde los elementos son otras listas.
* Ejemplos:
    1. suma_maximos([[1, 2, 3, 4], [5, 6, 7, 8], [20, 21, 34, 10]]) ➞ 46
    2. suma_maximos([[8, 9, -3], [3, 8, 4], [0, -2, 4]]) ➞ 21


````python
# Método 1: con comprensión de lista y sum()
def suma_maximos(matriz):
    return sum(max(fila) for fila in matriz)
````

````python
# Método 2: con un bucle anidado, recorriendo cada fila manualmente
def suma_maximos(matriz):
    total = 0
    for fila in matriz:
        maximo_fila = fila[0]
        for valor in fila:
            if valor > maximo_fila:
                maximo_fila = valor
        total += maximo_fila
    return total
````

````python
print(suma_maximos([[1, 2, 3, 4], [5, 6, 7, 8], [20, 21, 34, 10]]))
print(suma_maximos([[8, 9, -3], [3, 8, 4], [0, -2, 4]]))
````

### Comentario
- Una **matriz** en Python no es más que una lista cuyos elementos son, a su vez, otras listas (una "lista de listas"): cada lista interior representa una fila.
- `max(fila)` calcula el máximo de una única fila; `max(fila) for fila in matriz` es una expresión generadora que calcula ese máximo para **cada fila**, una tras otra; y `sum(...)` los suma todos.
- El Método 2 hace exactamente lo mismo, pero de forma más explícita con un bucle anidado: un bucle exterior que recorre cada fila, y dentro, un segundo bucle que calcula el máximo de esa fila comparando valor a valor (como en el Resuelto 20 del Nivel 1).

## Resuelto 151: Recreando la función `len()` para un carrito
* Cree una función que devuelva la cantidad de elementos de una cadena, de una lista o de un diccionario, sin usar `len()`.
* Ejemplos:
    1. longitud_manual("Hola Mundo") ➞ 10
    2. longitud_manual([2, 3, 4, -7]) ➞ 4
    3. longitud_manual({'a': 1, 'b': 2, 'c': 3}) ➞ 3
    4. longitud_manual({'uno', 1, '1'}) ➞ 3


````python
# Método 1: con un bucle contador
def longitud_manual(objeto):
    contador = 0
    for _ in objeto:
        contador += 1
    return contador
````

````python
print(longitud_manual("Hola Mundo"))
print(longitud_manual([2, 3, 4, -7]))
print(longitud_manual({'a': 1, 'b': 2, 'c': 3}))
print(longitud_manual({'uno', 1, '1'}))
````

- `for _ in objeto:` funciona igual para cadenas, listas, conjuntos o diccionarios, porque todos ellos son **iterables**: se pueden recorrer elemento a elemento con un bucle `for`.
- Un detalle importante: al recorrer un diccionario directamente con `for _ in diccionario:`, Python itera sobre sus **claves**, no sobre sus valores ni sobre pares clave-valor; por eso, en este ejercicio, contar las vueltas del bucle sobre un diccionario cuenta su número de claves (que coincide con su número de entradas).

## Resuelto 173: Recursividad aplicada a repetir un carácter `n` veces
* Cree una función recursiva que tome dos parámetros y repita un carácter `n` veces.
* El primer parámetro es el carácter que se repetirá.
* El segundo parámetro es el número de veces que se repetirá.
* Ejemplos:
    1. redoble("R", 3) ➞ "RRR"
    2. redoble("¡", 1) ➞ "¡"
    3. redoble("Ta", 2) ➞ "TaTa"


````python
# Método 1: recursivamente: caso base n==0, caso recursivo n-1
def redoble(caracter, n):
    if n == 0:
        return ""
    return caracter + redoble(caracter, n - 1)
````

````python
print(redoble("R", 3))
print(redoble("¡", 1))
print(redoble("Ta", 2))
````

### Comentario
- Esta es la primera función **recursiva** del curso: una función que se llama **a sí misma** para resolver un problema más pequeño, hasta llegar a un caso tan simple que ya no necesita seguir llamándose (el **caso base**).

```python
def redoble(caracter, n):
    if n == 0:              # caso base: ya no hay que repetir nada más
        return ""
    return caracter + redoble(caracter, n - 1)   # caso recursivo
```

- El **caso base** (`n == 0`) es imprescindible: sin él, la función se llamaría a sí misma indefinidamente (de forma parecida a un bucle infinito), hasta agotar la memoria disponible para las llamadas (un error `RecursionError`).
- El **caso recursivo** resuelve el problema apoyándose en una versión más pequeña del mismo problema: `redoble(caracter, n)` se construye a partir de `redoble(caracter, n - 1)`, acercándose un paso más al caso base en cada llamada.
- Por ejemplo, `redoble("R", 3)` internamente calcula `"R" + redoble("R", 2)`, que a su vez calcula `"R" + redoble("R", 1)`, que calcula `"R" + redoble("R", 0)`, que devuelve `""` (el caso base). Al "deshacer" esas llamadas de dentro hacia afuera, se obtiene `"R" + "R" + "R" + "" = "RRR"`.
- Cualquier función recursiva se podría escribir también con un bucle (de hecho, este mismo ejercicio se podría resolver con `caracter * n`, como en el Resuelto 58 del Nivel 1); la recursividad es, sobre todo, una forma distinta y muy potente de pensar ciertos problemas, que iremos viendo en más ejercicios.

## Resuelto 175: Semáforo intermitente
* Para este desafío no se da enunciado.
* Cree una función que responda a los casos de uso con la salida que se indica.
* Ejemplos:
    1. semaforo(['stop', 'avanza', 'stop', 'avanza', 'stop']) ➞ "o/o/o"
    2. semaforo(['stop', 'avanza', 'stop', 'stop', 'stop']) ➞ "o/o-o"
    3. semaforo(['avanza', 'avanza', 'avanza', 'avanza', 'avanza']) ➞ "|/|/|"
    4. semaforo(['stop', 'stop', 'stop', 'stop', 'stop']) ➞ "o-o-o"


````python
# Método 1: las posiciones pares son estados, las impares determinan el separador
def semaforo(acciones):
    simbolos = {"stop": "o", "avanza": "|"}
    estados = acciones[::2]
    conectores = acciones[1::2]
    resultado = simbolos[estados[0]]
    for conector, estado in zip(conectores, estados[1:]):
        resultado += ("/" if conector == "avanza" else "-") + simbolos[estado]
    return resultado
````

````python
print(semaforo(['stop', 'avanza', 'stop', 'avanza', 'stop']))
print(semaforo(['stop', 'avanza', 'stop', 'stop', 'stop']))
print(semaforo(['avanza', 'avanza', 'avanza', 'avanza', 'avanza']))
print(semaforo(['stop', 'stop', 'stop', 'stop', 'stop']))
````

### Comentario
- `acciones[::2]` y `acciones[1::2]` son ejemplos de slicing con **paso 2**: `[::2]` coge los elementos en las posiciones pares (0, 2, 4, ...) y `[1::2]` coge los de las posiciones impares (1, 3, 5, ...). Es una generalización del slicing con paso que ya vimos con `[::-1]` (Resuelto 43 del Nivel 1), pero aquí con un paso positivo distinto de 1.
- `zip(conectores, estados[1:])` recorre ambas listas en paralelo, tal y como vimos en el Resuelto 138, para emparejar cada conector con el estado que le sigue.

## Resuelto 182: Separar un número en dos partes iguales
* Dado un número `n`, devuelva una lista que contenga las dos mitades del número.
* Si el número es impar, haga que el mayor sea el que está más a la izquierda (al revés que en el Reto original).
* El número `n` siempre será un número entero.
* El número `n` puede ser negativo.
* Ejemplos:
    1. separar_numero_v2(4) ➞ [2, 2]
    2. separar_numero_v2(10) ➞ [5, 5]
    3. separar_numero_v2(11) ➞ [6, 5]
    4. separar_numero_v2(-9) ➞ [-4, -5]
    5. separar_numero_v2(0) ➞ [0, 0]


````python
# Método 1: usando división entera y ajustando para impares y negativos
def separar_numero_v2(n):
    mitad = abs(n) // 2
    if n >= 0:
        if n % 2 == 0:
            return [mitad, mitad]
        return [mitad + 1, mitad]
    else:
        if n % 2 == 0:
            return [-mitad, -mitad]
        return [-mitad, -(mitad + 1)]
````

````python
# Método 2: aprovechando la aritmética de Python con negativos
def separar_numero_v2(n):
    return [n // 2 + n % 2, n // 2]
````

````python
print(separar_numero_v2(4))
print(separar_numero_v2(10))
print(separar_numero_v2(11))
print(separar_numero_v2(-9))
print(separar_numero_v2(0))
````

### Comentario
- Este ejercicio ilustra una particularidad de la división entera `//` en Python con números **negativos**: a diferencia de otros lenguajes (donde `//` suele truncar hacia cero), en Python `//` siempre redondea **hacia menos infinito**. Por ejemplo, `-9 // 2` da `-5`, no `-4`.
- Por eso el Método 1 usa `abs(n) // 2` (calculando primero el valor absoluto) y solo al final vuelve a aplicar el signo negativo si `n` era negativo; así evita la asimetría de `//` con números negativos y controla manualmente qué mitad debe ser la mayor.
- El Método 2 consigue el mismo resultado en una sola línea, apoyándose en otra propiedad de Python: el resto `%` siempre tiene el mismo signo que el divisor (nunca el del dividendo), así que `n % 2` da siempre `0` o `1`, nunca `-1`, incluso con `n` negativo. Eso hace que `n // 2 + n % 2` "redondee hacia arriba" de forma consistente en todos los casos, positivos y negativos, sin necesidad de distinguir ramas por separado.
- En otros lenguajes, -9 // 2 = -4 (truncando hacia cero), pero en Python
    - -9 // 2 = -5 (redondeando hacia -infinito). Esta diferencia es clave:
        * -9 // 2 = -5
        * -9 % 2  = 1    (porque -9 = (-5)*2 + 1)
- Por eso, para n=-9: [-5 + 1, -5] = [-4, -5], que es exactamente lo que necesitamos: el mayor (-4) a la izquierda y el menor (-5) a la derecha.

````python
# Método 1: con comprensión de lista
def ensordinar(lst):
    return [n if n.endswith("m") else n + "m" for n in lst]
````

## Resuelto 194: Eliminar ceros al inicio y al final
* Cree una función que reciba una cadena de dígitos y retorne esa misma cadena pero habiendo eliminado los ceros al inicio y al final.
* Consiga esto sin utilizar la función `strip`.
* Ejemplos:
    1. eliminar_ceros("00123400") ➞ "1234"
    2. eliminar_ceros("0007890") ➞ "789"
    3. eliminar_ceros("5") ➞ "5"


````python
# Método 1: encontrando primer y último dígito no cero con bucles
def eliminar_ceros(cadena):
    inicio = 0
    while inicio < len(cadena) and cadena[inicio] == "0":
        inicio += 1
    fin = len(cadena) - 1
    while fin >= 0 and cadena[fin] == "0":
        fin -= 1
    return cadena[inicio:fin+1]
````

````python
print(eliminar_ceros("00123400"))
print(eliminar_ceros("0007890"))
print(eliminar_ceros("5"))
````

### Comentario
- Este ejercicio pide explícitamente **no usar `.strip()`**, aunque `cadena.strip("0")` resolvería el problema en una sola línea eliminando los ceros de ambos extremos. El objetivo es practicar cómo se recorren los extremos de una secuencia "a mano", con bucles `while`, antes de apoyarte en un método ya hecho.
- Los dos bucles `while` avanzan `inicio` desde el principio y retroceden `fin` desde el final, mientras encuentren un carácter `"0"`, hasta detenerse justo en el primer y el último carácter que no lo son. Ambos bucles comprueban también los límites (`inicio < len(cadena)` y `fin >= 0`) para no salirse de la cadena si todos los caracteres fueran `"0"`.

## Resuelto 196: Producto que aparece un número impar de veces
* Cree una función que tome una lista de códigos de producto (números) y encuentre el que aparece un número impar de veces.
* Siempre habrá un solo código que aparezca un número impar de veces.
* Ejemplos:
    1. codigo_impar([1, 1, 2, -2, 5, 2, 4, 4, -1, -2, 5]) ➞ -1
    2. codigo_impar([20, 1, 1, 2, 2, 3, 3, 5, 5, 4, 20, 4, 5]) ➞ 5
    3. codigo_impar([10]) ➞ 10


````python
# Método 1: con count() y un bucle
def codigo_impar(lista):
    for x in lista:
        if lista.count(x) % 2 == 1:
            return x
````

````python
# Método 2: con XOR (los pares se cancelan entre sí)
from functools import reduce
import operator

def codigo_impar(lista):
    return reduce(operator.xor, lista)
````

````python
print(codigo_impar([1, 1, 2, -2, 5, 2, 4, 4, -1, -2, 5]))
print(codigo_impar([20, 1, 1, 2, 2, 3, 3, 5, 5, 4, 20, 4, 5]))
print(codigo_impar([10]))
````

### Comentario
- El Método 1 es sencillo de entender, pero recalcula `lista.count(x)` para cada elemento, lo que resulta lento en listas muy grandes.
- El Método 2 usa un truco más avanzado basado en el operador `^` (**XOR**, "o exclusivo"): al aplicar XOR entre dos números iguales el resultado es `0`, y al aplicar XOR con `0` un número se queda igual. Por eso, si todos los números aparecen un número **par** de veces excepto uno, al combinarlos todos con XOR, los que se repiten un número par de veces "se cancelan" entre sí, y solo queda el que aparece un número impar de veces.
- `from functools import reduce` importa una función que aplica una operación (aquí, `operator.xor`) acumulativamente sobre todos los elementos de una lista, de dos en dos, hasta reducirla a un único valor: `reduce(operator.xor, [a, b, c])` equivale a `((a ^ b) ^ c)`.
- Este segundo método es mucho más eficiente que el primero, pero también más difícil de entender a simple vista; es un buen ejemplo de que, en programación, a veces hay que elegir entre un código más claro (Método 1) o uno más rápido pero menos intuitivo (Método 2).

## Resuelto 197: Filtrar números negativos de un array
* Cree una función que tome una lista de cadenas y números y devuelva una nueva lista solo con los números negativos (sin cadenas y sin números positivos ni cero).
* Los números de la lista no deben repetirse.
* Se debe mantener el orden original.
* Ejemplos:
    1. filtrar_negativos([1, -2, 'a', 'b']) ➞ [-2]
    2. filtrar_negativos([1, 'a', 'b', 0, -15, -15]) ➞ [-15]
    3. filtrar_negativos([1, 2, 'aasf', '-1', '-123', -123]) ➞ [-123]
    4. filtrar_negativos([5, '5', 'bye', '7', '77']) ➞ []


````python
# Método 1: filtrando negativos, eliminando duplicados con set
def filtrar_negativos(lista):
    vistos = set()
    resultado = []
    for x in lista:
        if isinstance(x, int) and not isinstance(x, bool) and x < 0:
            if x not in vistos:
                vistos.add(x)
                resultado.append(x)
    return resultado
````

````python
print(filtrar_negativos([1, -2, 'a', 'b']))
print(filtrar_negativos([1, 'a', 'b', 0, -15, -15]))
print(filtrar_negativos([1, 2, 'aasf', '-1', '-123', -123]))
print(filtrar_negativos([5, '5', 'bye', '7', '77']))
````

### Comentario
- `vistos` es un conjunto (`set`) que se usa para llevar la cuenta de los números que ya se han añadido al resultado, y así evitar duplicados.
- Fíjate en que el resultado se sigue construyendo como una **lista** (`resultado`), no como un `set`: esto es importante porque los conjuntos no mantienen ningún orden, y el enunciado pide conservar el orden original de aparición. El patrón "usar un `set` aparte solo para comprobar duplicados, pero guardar el resultado en una lista" es una forma habitual de eliminar duplicados sin perder el orden.

## Resuelto 199: Combinaciones de un candado
* Cree una función que tome un número entero, no negativo `n` y devuelva de cuántas formas distintas se pueden ordenar `n` objetos distintos en fila.

* Suponga que todas las entradas son mayores o iguales a 0.
* Ejemplos:

    4. formas_de_ordenar(0) ➞ 1

````python
# Método 2: con math.factorial()
import math

def formas_de_ordenar(n):
    return math.factorial(n)
````

````python
# Método 3: de forma recursiva
def formas_de_ordenar(n):
    if n == 0:
        return 1
    return n * formas_de_ordenar(n - 1)
````

## Resuelto 200: Potencia por recursión
* Escriba una función que calcule `base` elevado a `exponente` de forma recursiva, sin usar el operador `**`.
* El exponente siempre será un número entero mayor o igual a 0.
* Cualquier número elevado a 0 da 1, por definición.
* Ejemplos:
    1. potencia_recursiva(2, 5) ➞ 32
    2. potencia_recursiva(3, 3) ➞ 27
    3. potencia_recursiva(5, 0) ➞ 1
    4. potencia_recursiva(10, 2) ➞ 100


````python
# Método 1: recursivamente: caso base exponente==0, caso recursivo base * potencia(base, exp-1)
def potencia_recursiva(base, exponente):
    if exponente == 0:
        return 1
    return base * potencia_recursiva(base, exponente - 1)
````

````python
print(potencia_recursiva(2, 5))
print(potencia_recursiva(3, 3))
print(potencia_recursiva(5, 0))
print(potencia_recursiva(10, 2))
````

### Comentario
- Esta es la tercera función recursiva del curso (después de los Resueltos 173 y 199), y sigue exactamente el mismo patrón: un **caso base** (`exponente == 0`, que da `1`, ya que cualquier número elevado a 0 es 1) y un **caso recursivo** que multiplica `base` por el resultado de `potencia_recursiva` con un exponente una unidad menor.
- Fíjate en el paralelismo con el factorial del Resuelto 199: allí se iba multiplicando y **restando 1** al número en cada llamada; aquí se multiplica y también se **resta 1** al exponente. Es el mismo patrón recursivo aplicado a un problema distinto.
