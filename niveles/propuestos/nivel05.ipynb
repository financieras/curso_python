## Resuelto 401: Relleno con ceros, asteriscos y guiones
* Para este reto no necesitas crear una función.
* En su lugar, necesitas elaborar una cadena de plantilla que pueda ser formateada para obtener un cierto resultado.
* Escribe tres cadenas de plantilla según el ejemplo, cada una usando un carácter y una dirección de relleno distintos a los del Reto 401.
* Ejemplo:
```python
ceros = "tucadenadeplantillaaquí"
asteriscos = "tucadenadeplantillaaquí"
guion_bajo = "tucadenadeplantillaaquí"
ceros.format("Cero") # "0000000000000000Cero"
asteriscos.format("Estrella") # "****Estrella*****"
guion_bajo.format("Uno") # "Uno________________"
```
* Todas las cadenas finales deben tener una longitud de 20 caracteres.
* Envía una cadena, no una función.

````python
ceros = "{:0>20}"
asteriscos = "{:*^20}"
guion_bajo = "{:_<20}"
````

````python
print(ceros.format("Cero"))
print(asteriscos.format("Estrella"))
print(guion_bajo.format("Uno"))
````

### Comentario
- Estas cadenas usan la **mini-sintaxis de formato** de Python dentro de `{...}`, con la forma general `{:relleno alineación ancho}`:
  - `{:0>20}` → rellena con `0`, alineado a la derecha (`>`), hasta 20 caracteres de ancho
  - `{:*^20}` → rellena con `*`, centrado (`^`), hasta 20 caracteres
  - `{:_<20}` → rellena con `_`, alineado a la izquierda (`<`), hasta 20 caracteres
- El carácter de relleno se escribe justo antes del símbolo de alineación (`>`, `^` o `<`), y después va el ancho total deseado.
- Estas plantillas son cadenas normales que aún no se han "rellenado": `.format(...)` es lo que sustituye el hueco `{}` por el valor indicado, aplicando el formato descrito.

## Resuelto 402: Vueltas de un tiovivo
* Dada una lista de direcciones de giro de un tiovivo, "horario" o "antihorario", devuelve un número entero que representa cuántas vueltas completas de 360° se realizaron.
* Cada palabra en la lista cuenta como un giro de 90° en esa dirección.
* Devuelva un número entero, no negativo.
* Ejemplos:
    1. vueltas_360(['horario', 'horario', 'horario', 'horario', 'antihorario', 'horario']) ➞ 1
    2. vueltas_360(['antihorario', 'horario', 'antihorario', 'horario']) ➞ 0
    3. vueltas_360(['horario', 'horario', 'horario', 'horario', 'horario', 'horario', 'horario', 'horario']) ➞ 2
    4. vueltas_360(['antihorario', 'antihorario', 'antihorario', 'antihorario']) ➞ 1

````python
# Método 1: contando giros netos
import math
def vueltas_360(direcciones):
    grados = sum(90 if d == "horario" else -90 for d in direcciones)
    return abs(grados) // 360
````

````python
# Método 2: con un contador
def vueltas_360(direcciones):
    horario = direcciones.count("horario")
    antihorario = direcciones.count("antihorario")
    return abs(horario - antihorario) // 4
````

````python
print(vueltas_360(['horario', 'horario', 'horario', 'horario', 'antihorario', 'horario']))
print(vueltas_360(['antihorario', 'horario', 'antihorario', 'horario']))
print(vueltas_360(['horario', 'horario', 'horario', 'horario', 'horario', 'horario', 'horario', 'horario']))
print(vueltas_360(['antihorario', 'antihorario', 'antihorario', 'antihorario']))
````

## Resuelto 403: Cambiar Ingredientes Mixtos
* Cree una función que cambie todos los elementos en una lista (los ingredientes de una receta) de la siguiente manera:
1. Suma 1 a todos los enteros pares, no hace nada con los enteros impares.
2. Concatena "!" a todas las cadenas y pone en mayúscula la primera letra.
3. Cambia todos los valores booleanos por su opuesto.
* No habrá valores flotantes.
* No obtendrás cadenas con números y letras al mismo tiempo.
* Ejemplos:
    1. cambiar_ingredientes(['sal', 4, True]) ➞ ['Sal!', 5, False]
    2. cambiar_ingredientes([7, 'sal', '3', 'harina']) ➞ [7, 'Sal!', '3!', 'Harina!']
    3. cambiar_ingredientes([False, 'False', 'true']) ➞ [True, 'False!', 'True!']

````python
# Método 1: procesando según el tipo
import math
def cambiar_ingredientes(lista):
    resultado = []
    for x in lista:
        if isinstance(x, bool):
            resultado.append(not x)
        elif isinstance(x, int) and not isinstance(x, bool):
            resultado.append(x + 1 if x % 2 == 0 else x)
        elif isinstance(x, str):
            resultado.append(x.capitalize() + "!")
        else:
            resultado.append(x)
    return resultado
````

````python
print(cambiar_ingredientes(['sal', 4, True]))
print(cambiar_ingredientes([7, 'sal', '3', 'harina']))
print(cambiar_ingredientes([False, 'False', 'true']))
````

* Su código muta la lista original por error.

* Código erróneo:

```python
def reducir_marcador(x, n):
    for i in range(len(x)):
        j = 1
        while j <= n:
            if i >= j and i < len(x) - j:
                x[i] -= 1
            j += 1
    return x
```

* Ejemplos:
    1. reducir_marcador([5, 5, 5, 5, 5], 0) ➞ [5, 5, 5, 5, 5]
    2. reducir_marcador([5, 5, 5, 5, 5], 1) ➞ [5, 4, 4, 4, 5]
    3. reducir_marcador([5, 5, 5, 5, 5], 2) ➞ [5, 4, 3, 4, 5]

````python
# Método 1: trabajando sobre una copia para no mutar el original
def reducir_marcador(x, n):
    copia = x.copy()
    for i in range(len(copia)):
        for j in range(1, n + 1):
            if i >= j and i < len(copia) - j:
                copia[i] -= 1
    return copia
````

## Resuelto 408: Contraseña Parcialmente Oculta
* Crea una función que tome una frase (varias contraseñas separadas por espacio) y transforme cada palabra utilizando las siguientes reglas:
1. Mantén el primer y último carácter igual.
2. Transforma los caracteres intermedios en un asterisco *.
* Las palabras con dos o menos letras no deben ser ocultadas en absoluto.
* Ejemplos:
    1. ocultar_contrasena("Aprender Python") ➞ "A******r P****n"
    2. ocultar_contrasena("La masa es energía") ➞ "La m**a es e*****a"
    3. ocultar_contrasena("Saltar el charco") ➞ "S****r el c****o"

````python
# Método 1: con split() y join()
def ocultar_contrasena(frase):
    def ocultar(p):
        if len(p) <= 2:
            return p
        return p[0] + "*" * (len(p) - 2) + p[-1]
    return " ".join(ocultar(p) for p in frase.split())
````

````python
print(ocultar_contrasena("Aprender Python"))
print(ocultar_contrasena("La masa es energía"))
print(ocultar_contrasena("Saltar el charco"))
````

## Resuelto 411: Arcoseno, Arcocoseno y Arcotangente
* Crea tres funciones que hagan tres cosas:
1. Multiplicar un número por el arcoseno de otro número.
2. Multiplicar un número por el arcocoseno de otro número.
3. Multiplicar un número por la arcotangente de otro número.
* En cada función, se te darán 2 números: x e y, donde y está entre -1 y 1 (para arcoseno y arcocoseno).
* El resultado estará en grados, no en radianes.
* Redondea el resultado a 2 lugares decimales.
* Ejemplos:
    1. arcoseno(8, 0.45) ➞ 8 × asin(0.45) en grados ➞ 213.95
    2. arcocoseno(10, 0.5) ➞ 10 × acos(0.5) en grados ➞ 600.0
    3. arcotangente(4, 1) ➞ 4 × atan(1) en grados ➞ 180.0

````python
# Método 1: usando math.asin, math.acos, math.atan
import math
def arcoseno(x, y):
    return round(x * math.degrees(math.asin(y)), 2)

def arcocoseno(x, y):
    return round(x * math.degrees(math.acos(y)), 2)

def arcotangente(x, y):
    return round(x * math.degrees(math.atan(y)), 2)
````

### Comentario
- `math.asin()`, `math.acos()` y `math.atan()` son las funciones trigonométricas **inversas** del seno, coseno y tangente: dado un valor, devuelven el ángulo (en radianes) que produce ese valor.
- Como el resultado de estas funciones siempre viene en **radianes**, `math.degrees(...)` se usa para convertirlo a grados, que es la unidad que pide el enunciado. La función inversa, `math.radians(...)`, haría la conversión contraria (de grados a radianes).

## Resuelto 412: Anidación de Diccionarios
* Cree una función que devuelva el valor original de un diccionario con una gran cantidad de sub-diccionarios anidados, todos con la misma clave "x".
* Solo necesita recuperar un elemento.
* Ejemplos:
    1. de_nest_dict({'x': {'x': {'x': {'x': 3}}}}) ➞ 3
    2. de_nest_dict({'x': {'x': {'x': {'x': {'x': {'x': True}}}}}}) ➞ True
    3. de_nest_dict({'x': {'x': 'bit'}}) ➞ "bit"

````python
# Método 1: con un bucle while
import math
def de_nest_dict(d):
    while isinstance(d, dict):
        d = d["x"]
    return d
````

````python
print(de_nest_dict({'x': {'x': {'x': {'x': 3}}}}))
print(de_nest_dict({'x': {'x': {'x': {'x': {'x': {'x': True}}}}}}))
print(de_nest_dict({'x': {'x': 'bit'}}))
````

### Comentario
- `while isinstance(d, dict):` es una forma de "ir bajando" por una estructura anidada sin saber de antemano cuántos niveles tiene: mientras el valor actual siga siendo un diccionario, se reemplaza por su valor asociado a `"x"`, y el bucle se repite. En cuanto `d` deja de ser un diccionario (es un número, un booleano, una cadena...), el bucle se detiene y ese es el valor final que se devuelve.

## Resuelto 414: Listas alternadas
* Una lista es alternada si cada índice par contiene un número negativo y cada índice impar contiene un número positivo.
* Crea una función que devuelva `True` si un arreglo es alternado, y `False` en caso contrario.
* Ejemplos:
    1. es_lista_alternada([-2, 7, -4, 9, -6, 1, -6, 3]) ➞ True
    2. es_lista_alternada([-2, 7, 9, 1, -6, 1, -6, 3]) ➞ False
    3. es_lista_alternada([-2, 7, -8, -8, -6, 1, -6, 3]) ➞ False

````python
# Método 1: comprobando par/impar
import math
def es_lista_alternada(lista):
    for i, x in enumerate(lista):
        if i % 2 == 0 and x >= 0:
            return False
        if i % 2 != 0 and x < 0:
            return False
    return True
````

## Resuelto 415: Transformar Reproducciones
* Cree una función que transforme una cadena de recuentos de reproducciones de un vídeo en una lista de números.
* Cada "M" representa un millón.
* Devuelve las reproducciones como una lista.
* Ejemplos:
    1. transform_plays("6.8M 13.5M") ➞ [6800000, 13500000]

    3. transform_plays("20.3M 3.8M 7.7M 992") ➞ [20300000, 3800000, 7700000, 992]

````python
# Método 1: procesando cada token
import math
def transform_plays(cadena):
    resultado = []
    for token in cadena.split():
        if token.endswith("M"):
            resultado.append(int(float(token[:-1]) * 1000000))
        else:
            resultado.append(int(token))
    return resultado
````

## Resuelto 416: Mínimo Común Múltiplo de Dos Números
* Escriba una función que devuelva el mínimo común múltiplo (mcm) de dos enteros.
* Ambos valores serán positivos.
* El mcm es el menor múltiplo positivo que es divisible por ambos números.
* Ejemplos:
    1. mcm(4, 6) ➞ 12
    2. mcm(21, 6) ➞ 42
    3. mcm(17, 13) ➞ 221
    4. mcm(4, 5) ➞ 20

````python
# Método 1: con el algoritmo de Euclides para MCD
import math
def mcm(a, b):
    def mcd(x, y):
        while y:
            x, y = y, x % y
        return x
    return abs(a * b) // mcd(a, b)
````

````python
# Método 2: con math.gcd()
import math
def mcm(a, b):
    return abs(a * b) // math.gcd(a, b)
````

````python
print(mcm(4, 6))
print(mcm(21, 6))
print(mcm(17, 13))
print(mcm(4, 5))
````

### Comentario
- El **máximo común divisor** (MCD) de dos números es el número más grande que divide a ambos exactamente. El Método 1 lo calcula con el **algoritmo de Euclides**: mientras `y` no sea cero, se sustituyen `x` e `y` por `y` y el resto de `x` entre `y` (`x % y`); cuando `y` llega a 0, `x` es el MCD. Es uno de los algoritmos más antiguos que se conocen (más de 2000 años) y también uno de los más eficientes.
- Una vez se tiene el MCD, el **mínimo común múltiplo** (MCM) se calcula con la fórmula `abs(a * b) // mcd(a, b)`, sin necesidad de probar múltiplos uno a uno.
- El Método 2 usa `math.gcd(a, b)`, que ya viene incluida en Python y hace exactamente el mismo cálculo que la función `mcd()` escrita a mano en el Método 1.

## Resuelto 417: Recursión: Números Triangulares
* Los números triangulares se crean de la siguiente manera:
- T(0) = 0
- T(n) = T(n-1) + n
* Escriba una función que calcule el n-ésimo número triangular, usando recursión.
* Ejemplos:
    1. triangular(0) ➞ 0
    2. triangular(1) ➞ 1
    3. triangular(5) ➞ 15
    4. triangular(8) ➞ 36
    5. triangular(20) ➞ 210

````python
# Método 1: recursivamente
import math
def triangular(n):
    if n == 0:
        return 0
    return n + triangular(n - 1)
````

````python
# Método 2: con la fórmula cerrada n*(n+1)//2
def triangular(n):
    return n * (n + 1) // 2
````

````python
print(triangular(0))
print(triangular(1))
print(triangular(5))
print(triangular(8))
print(triangular(20))
````

## Resuelto 427: La Conjetura de Collatz con lista de pasos
* Considere la misma operación de Collatz que en el Reto 427.
* Cree una función que devuelva la lista completa de números por los que pasa la secuencia hasta llegar a 1 (incluyendo el número inicial y el 1 final).
* Ejemplos:
    1. collatz_secuencia(2) ➞ [2, 1]
    2. collatz_secuencia(3) ➞ [3, 10, 5, 16, 8, 4, 2, 1]
    3. collatz_secuencia(10) ➞ [10, 5, 16, 8, 4, 2, 1]

````python
# Método 1: con un bucle while
import math
def collatz_secuencia(n):
    resultado = [n]
    while n != 1:
        n = n // 2 if n % 2 == 0 else n * 3 + 1
        resultado.append(n)
    return resultado
````

````python
# Método 2: recursivamente
def collatz_secuencia(n):
    if n == 1:
        return [1]
    sig = n // 2 if n % 2 == 0 else n * 3 + 1
    return [n] + collatz_secuencia(sig)
````

````python
print(collatz_secuencia(2))
print(collatz_secuencia(3))
print(collatz_secuencia(10))
````

### Comentario
- La conjetura de Collatz (también llamada 3n+1) dice que, partiendo de cualquier número entero positivo y aplicando repetidamente "si es par, divide entre 2; si es impar, multiplica por 3 y suma 1", siempre se acaba llegando a 1. Es una conjetura, no un teorema: se ha comprobado para muchísimos números, pero nadie ha demostrado todavía que sea cierta para **todos** los números, así que, en teoría, ambos métodos (`while` y recursivo) podrían no terminar nunca para algún número aún no probado (aunque en la práctica siempre termina para los números con los que se suele trabajar).

## Resuelto 435: Factorial Primorial
* El primorial de un número primo `p` (denotado `p#`) es el producto de todos los números primos menores o iguales a `p`.
* Ejemplo: 5# = 2 × 3 × 5 = 30
* Cree una función que tome un número `p` y devuelva su primorial, asumiendo que siempre se da un número primo.
* Intenta resolverlo con recursión.
* Ejemplos:
    1. primorial(2) ➞ 2
    2. primorial(3) ➞ 6
    3. primorial(5) ➞ 30
    4. primorial(7) ➞ 210
    5. primorial(11) ➞ 2310

````python
# Método 1: recursivamente buscando el primo anterior
import math
def primorial(p):
    def es_primo(n):
        if n < 2: return False
        for i in range(2, int(n**0.5)+1):
            if n % i == 0: return False
        return True
    if p < 2:
        return 1
    return p * primorial(p - 1) if es_primo(p) else primorial(p - 1)
````

````python
# Método 2: iterativamente
def primorial(p):
    def es_primo(n):
        if n < 2: return False
        for i in range(2, int(n**0.5)+1):
            if n % i == 0: return False
        return True
    resultado = 1
    for i in range(2, p + 1):
        if es_primo(i):
            resultado *= i
    return resultado
````

    3. unique_in_order_count([1, 2, 2, 3, 3]) ➞ [(1, 1), (2, 2), (3, 2)]

````python
# Método 1: contando rachas consecutivas
import math
def unique_in_order_count(sequence):
    resultado = []
    actual = None
    count = 0
    for x in sequence:
        if x == actual:
            count += 1
        else:
            if actual is not None:
                resultado.append((actual, count))
            actual = x
            count = 1
    if actual is not None:
        resultado.append((actual, count))
    return resultado
````

## Resuelto 458: ¿Cuántas Raíces Enteras Tiene Esta Cuadrática?
* Una ecuación cuadrática ax² + bx + c = 0 puede tener 0, 1 o 2 soluciones reales.
* Dados a, b y c, debes devolver cuántas de esas soluciones (si existen) son números enteros.
* `a` siempre será distinto de cero.
* Ejemplos:
    1. raices_enteras(1, 0, -4) ➞ 2
    2. raices_enteras(1, -3, 2) ➞ 2
    3. raices_enteras(1, 0, 1) ➞ 0
    4. raices_enteras(1, 1, 1) ➞ 0

````python
# Método 1: con la fórmula cuadrática
import math
def raices_enteras(a, b, c):
    discriminante = b**2 - 4*a*c
    if discriminante < 0:
        return 0
    r1 = (-b + math.sqrt(discriminante)) / (2*a)
    r2 = (-b - math.sqrt(discriminante)) / (2*a)
    count = 0
    if r1 == int(r1): count += 1
    if r2 == int(r2) and r1 != r2: count += 1
    return count
````

````python
print(raices_enteras(1, 0, -4))
print(raices_enteras(1, -3, 2))
print(raices_enteras(1, 0, 1))
print(raices_enteras(1, 1, 1))
````

* Crea una función que tome dos números como argumentos (num, length) y devuelva una lista de los primeros `length` divisores positivos de `num`, en orden ascendente.
* Si `num` tiene menos de `length` divisores, devuelve todos los que tenga.
* Ejemplos:
    1. list_of_divisors(12, 3) ➞ [1, 2, 3]
    2. list_of_divisors(100, 5) ➞ [1, 2, 4, 5, 10]
    3. list_of_divisors(7, 6) ➞ [1, 7]

````python
# Método 1: generando divisores
import math
def list_of_divisors(num, length):
    divisores = []
    for i in range(1, int(num**0.5) + 1):
        if num % i == 0:
            divisores.append(i)
            if i != num // i:
                divisores.append(num // i)
    divisores.sort()
    return divisores[:length]
````

* Dos líneas son perpendiculares si el producto de sus pendientes es -1. La pendiente de una línea ax+by=c es -a/b.
* Todos los casos de prueba usan entradas válidas.
* Todos los coeficientes serán números enteros.
* Ejemplos:
    1. lines_are_perpendicular([1, 1, 3], [1, -1, 4]) ➞ True

    3. lines_are_perpendicular([2, 1, 5], [1, -2, 1]) ➞ True

````python
# Método 1: producto de pendientes = -1
import math
def lines_are_perpendicular(line1, line2):
    a1, b1, _ = line1
    a2, b2, _ = line2
    return round((-a1/b1) * (-a2/b2), 5) == -1.0
````

* Considerar que en el caso de una lista vacía la suma es cero.
* Si el último número de la lista no es negativo, la suma será cero.
* Ejemplos:
    1. suma_ultimos_negativos([3, 5, -1, 2, 7, -3, -1, -4]) ➞ 8
    2. suma_ultimos_negativos([3, 5, 1, 2, 7, -3, -1, -4]) ➞ 8
    3. suma_ultimos_negativos([-1, 2, 7, -3, 1, 4]) ➞ 0
    4. suma_ultimos_negativos([]) ➞ 0

````python
# Método 1: recorriendo de derecha a izquierda
import math
def suma_ultimos_negativos(lista):
    total = 0
    for n in reversed(lista):
        if n < 0:
            total += -n
        else:
            break
    return total
````

## Resuelto 495: Frases invertidas en un texto
* Dado un texto de entrada (varias frases separadas por punto), invierta el texto frase por frase (la última frase pasa a ser la primera).
* Cada frase mantiene sus palabras intactas y en el mismo orden interno.
* El texto de entrada puede contener espacios adicionales alrededor de los puntos, pero el resultado no debe contener espacios iniciales o finales en cada frase.
* Ejemplos:
    1. reverse_frases("El cielo es azul. El mar es profundo. La luna brilla.") ➞ "La luna brilla. El mar es profundo. El cielo es azul."
    2. reverse_frases("Hola mundo.  Adios mundo.  ") ➞ "Adios mundo. Hola mundo."

````python
# Método 1: con split() y join()
import math
def reverse_frases(texto):
    frases = [f.strip() for f in texto.split(".") if f.strip()]
    return ". ".join(reversed(frases)) + "."
````

````python
print(reverse_frases("El cielo es azul. El mar es profundo. La luna brilla."))
print(reverse_frases("Hola mundo.  Adios mundo.  "))
````

## Resuelto 498: Constructor de Nuevos Códigos
* Crea una función que construya un código a partir de los dígitos desordenados contenidos en la primera lista.
* Usa la segunda lista para establecer cada posición de los dígitos en la primera lista.
* Devuelve una cadena de texto con los dígitos ordenados (que forman el código).
* Los elementos en la segunda lista son índices de los elementos en la primera lista.
* Ejemplos:
    1. code_builder(['5', '2', '9'], [1, 0, 2]) ➞ "259"
    2. code_builder(['7', '0', '1', '3'], [3, 0, 2, 1]) ➞ "0317"

````python
# Método 1: con sorted() por la segunda lista
import math
def code_builder(digitos, positions):
    pares = sorted(zip(positions, digitos))
    return "".join(d for _, d in pares)
````

````python
print(code_builder(['5', '2', '9'], [1, 0, 2]))
print(code_builder(['7', '0', '1', '3'], [3, 0, 2, 1]))
````

### Comentario
- `zip(positions, digitos)` empareja cada posición con su dígito correspondiente; `sorted(...)` sobre esas parejas las ordena por su primer elemento (la posición), reconstruyendo así el orden correcto sin necesidad de crear manualmente una lista vacía e ir colocando cada dígito en su sitio con índices.

    2. retroceder("czf") ➞ "bye"
    3. retroceder("xfmdpnf") ➞ "welcome"

````python
# Método 1: con chr() y ord()
import math
def retroceder(texto):
    return "".join(chr(ord(c) - 1) for c in texto)
````

## Resuelto 500: Función Histograma Vertical
* Construye una función que cree histogramas, pero esta vez de forma vertical (cada número es una columna, no una fila).
* El segundo argumento de la función representa el carácter que se debe utilizar.
* Para simplificar, todas las columnas tendrán la misma altura que el valor máximo de la lista, rellenando con espacios por arriba.
* Ejemplos:
    1. histograma_vertical([1, 3, 2], "#") ➞ " #
 ##
###"
    2. histograma_vertical([2, 4], "=") ➞ " =
 =
==
=="

````python
# Método 1: construyendo línea a línea
def histograma_vertical(lista, caracter):
    max_h = max(lista)
    lineas = []
    for nivel in range(max_h, 0, -1):
        linea = "".join(caracter if col >= nivel else " " for col in lista)
        lineas.append(linea)
    return "\n".join(lineas)
````

````python
# Método 2: con zip_longest
def histograma_vertical(lista, caracter):
    max_h = max(lista)
    columnas = []
    for col in lista:
        columnas.append(" " * (max_h - col) + caracter * col)
    return "\n".join("".join(fila) for fila in zip(*columnas))
````

````python
print(histograma_vertical([1, 3, 2], "#"))
print(histograma_vertical([2, 4], "="))
````

### Comentario
- El Método 2 usa `zip(*columnas)`, un truco muy característico de Python: el asterisco `*` delante de `columnas` **desempaqueta** la lista de columnas, pasando cada una como un argumento independiente a `zip()` (equivale a escribir `zip(columnas[0], columnas[1], columnas[2], ...)`, sin importar cuántas columnas haya).
- El resultado es que `zip()` agrupa el primer carácter de cada columna, luego el segundo, y así sucesivamente, lo que en la práctica **transpone** la lista de columnas: convierte columnas en filas. Este patrón, `zip(*lista_de_listas)`, es la forma más habitual en Python de "girar" una matriz representada como lista de listas.
- Un pequeño detalle: el comentario del código dice "con zip_longest", pero en realidad se usa `zip()` normal, no `itertools.zip_longest`. Aquí funciona bien porque todas las columnas se rellenan antes con espacios hasta la misma altura (`max_h`), así que todas tienen igual longitud y `zip()` normal es suficiente; `zip_longest` solo sería necesario si las columnas pudieran tener longitudes distintas.
